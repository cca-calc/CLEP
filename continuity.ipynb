## 1.   Core Definitions You Must Know

### **1.1 Definition of Continuity at a Point**
A function $ f $ is continuous at $ x = a $ if:

$
\lim_{x \to a} f(x) = f(a)
$

This requires three things:

- $ f(a) $ exists  
- $ \lim_{x \to a} f(x) $ exists 
- $ \lim_{x \to a} f(x) = f(a)$  


### **1.2 Types of Discontinuities**
Know how to identify each from graphs, tables, or formulas:
Remember the limit exists only if  $ \lim_{x \to a^-} f(x) = \lim_{x \to a^+} f(x) = f(a) $ (page 87)

| Type | Description | CLEP Nuance |
|------|-------------|-------------|
| **Removable** | Hole; limit exists but $ f(a) \neq \lim f(x) $ | CLEP loves asking if redefining a point can make a function continuous |
| **Jump** | Left and right limits exist but differ | Often tested with piecewise functions |
| **Infinite** | Vertical asymptote | Usually tied to rational functions |
| **Oscillatory** | Wild oscillation near a point | Rare but possible conceptual question |

####  Why a Limit Can Exist Even When There Is a Hole (Removable Discontinuity) (This is a convention of the CLEP)

A **hole discontinuity** occurs when the *behavior* of the function near a point is well‑defined and consistent, but the **function value at that point is missing or incorrect**.

The key idea:  
**A limit depends only on nearby values, not the value at the point itself.**

## 1. The Limit Looks at the Trend, Not the Actual Value

When we compute:

 $
\lim_{x \to a} f(x)
 $

we are asking:

> “As \( x \) gets arbitrarily close to \( a \), what value does \( f(x) \) approach?”

This question **does not require** that \( f(a) \) exists.

So even if:
- \( f(a) \) is undefined  
- or \( f(a) \) is defined incorrectly  
- or the graph has a hole at \( x = a \)

the limit can still exist **as long as the values of \( f(x) \) approach a single number** from both sides.

## 2.   Continuity of Common Function Classes

### **2.1 Always Continuous on Their Domains**
- Polynomials  
- Rational functions (except where denominator = 0)  
- Exponential and logarithmic functions  
- Trig functions (sin, cos, tan, etc.)  
- Root functions (continuous where defined)

CLEP frequently asks:  
> “Where is this function continuous?”  
Your job: identify domain restrictions.

---

## 3.   Continuity of Piecewise Functions

This is a **high‑yield CLEP topic**.

### What to check:
- Evaluate left‑hand limit  
- Evaluate right‑hand limit  
- Evaluate function value  
- Ensure all three match

CLEP nuance:  
They often give a piecewise function with a parameter $ k $ and ask you to **solve for $ k $** so the function is continuous.

---

## 4.   Intermediate Value Theorem (IVT)

### **4.1 Statement**
If $ f $ is continuous on $[a,b]$, then for any value $ N $ between $ f(a) $ and $ f(b) $, there exists $ c \in (a,b) $ such that:

 $
f(c) = N
$

### **4.2 CLEP Nuances**
- They test **existence**, not solving the equation.  
- They love sign‑change problems:  
  - If $ f(a) < 0 $ and $ f(b) > 0 $, then a root exists.  
- They may ask whether IVT applies when the interval is **not closed** or the function is **not continuous**.

---

## 5.   Extreme Value Theorem (EVT) and Continuity

### **5.1 Statement**
If $ f $ is continuous on a **closed** interval $[a,b]$, then it has both a max and a min.

### **5.2 CLEP Nuances**
- They test whether the hypotheses are satisfied.  
- They may give an open interval and ask whether EVT still applies (it doesn’t).  
- They may ask you to identify where extrema occur (endpoints vs critical points).

## 6.   Limits and Continuity

### **6.1 Limit Laws**
You must be fluent with:
- Direct substitution  
- Factoring  
- Rationalizing  
- One‑sided limits  
- Limits from graphs  

CLEP nuance:  
They often test whether a limit exists even when the function value is irrelevant.

### **6.2 Continuity via Limits**
You should be able to:
- Determine continuity from a limit  
- Identify discontinuities from limit mismatch  
- Evaluate limits of piecewise functions

---

## 7.   Continuity and Differentiability

### **7.1 Key Relationship**
If a function is differentiable at $ a $, it is continuous at $ a $.  
The converse is **not** true.

CLEP nuance:  
They love asking which graphs are differentiable vs continuous.  
Common nondifferentiable but continuous points:
- Corners  
- Cusps  
- Vertical tangents  

---

## 8.   Continuity in Applied Contexts (CLEP loves these)

### Common scenarios:
- Temperature changing over time  
- Height of water in a tank  
- Position of a moving object  
- Cost or revenue functions  

CLEP nuance:  
They test whether a model **must** be continuous (e.g., temperature) or **may not** be (e.g., cost jumps).

## 9.   High‑Yield CLEP Question Types

### **9.1 Identify where a function is continuous**
Usually rational or piecewise.

### **9.2 Solve for a constant to make a function continuous**
Classic piecewise continuity problem.

### **9.3 Apply IVT to guarantee a solution**
Often a sign‑change question.

### **9.4 Determine whether EVT applies**
Check continuity and closed interval.

### **9.5 Graph‑based continuity questions**
Identify holes, jumps, asymptotes.

### **9.6 Continuity vs differentiability**
Corners, cusps, vertical tangents.
