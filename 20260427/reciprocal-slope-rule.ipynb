# Deriving the Reciprocal Rule for Inverse Functions

We assume:

- $ f $ is differentiable
- $ f $ is one‑to‑one (so it has an inverse)
- $ g = f^{-1} $

We want to derive the rule:

$$
(f^{-1})'(y) = \frac{1}{f'(x)} \quad \text{where } y = f(x)
$$

Below is the derivation in **two parallel notations**:

- **Left column:** traditional inverse notation $ f^{-1} $
- **Right column:** using $ g(x) = f^{-1}(x) $

---

## Step 1 — Start with the identity

| Traditional notation | Using $ g(x) = f^{-1}(x) $ |
|----------------------|------------------------------|
| $ f(f^{-1}(x)) = x $ | $ f(g(x)) = x $ |

This identity holds because $ f^{-1} $ undoes $ f $.

---

## Step 2 — Differentiate both sides with respect to $ x $

Apply the chain rule.

| Traditional notation | Using $ g(x) = f^{-1}(x) $ |
|----------------------|------------------------------|
| $ f'(f^{-1}(x)) \cdot (f^{-1})'(x) = 1 $ | $ f'(g(x)) \cdot g'(x) = 1 $ |

---

## Step 3 — Solve for the derivative of the inverse

| Traditional notation | Using $ g(x) = f^{-1}(x) $ |
|----------------------|------------------------------|
| $ (f^{-1})'(x) = \dfrac{1}{f'(f^{-1}(x))} $ | $ g'(x) = \dfrac{1}{f'(g(x))} $ |

This is the **reciprocal rule**.

---

## Step 4 — Rewrite using corresponding points

Let $ y = f(x) $. Then:

- $ x = f^{-1}(y) $
- The slope of $ f^{-1} $ at $ y $ depends on the slope of $ f $ at the corresponding $ x $

| Traditional notation | Using $ g(x) = f^{-1}(x) $ |
|----------------------|------------------------------|
| $ (f^{-1})'(y) = \dfrac{1}{f'(x)} \quad \text{where } y = f(x) $ | $ g'(y) = \dfrac{1}{f'(x)} \quad \text{where } y = f(x) $ |

---

## Interpretation

- The point $ (x, y) $ lies on $ f $
- The point $ (y, x) $ lies on $ f^{-1} $
- Their slopes are reciprocals:

$$
\text{slope of inverse} = \frac{1}{\text{slope of original}}
$$

This happens because inverse functions **swap the axes**, and swapping rise/run produces a reciprocal.

---

## Example: $ f(x) = 4x + 2 $

For a linear function the inverse is easy to find by hand, making it a clean check of the reciprocal-slope rule.

**Find the inverse.**  Solve $ y = 4x + 2 $ for $ x $:

$$
x = \frac{y - 2}{4} \quad \Longrightarrow \quad f^{-1}(y) = \frac{1}{4}y - \frac{1}{2}
$$

**Apply the rule.**

| Quantity | Value |
|----------|-------|
| $ f'(x) $ | $ 4 $ (constant slope) |
| $ (f^{-1})'(y) $ | $ \dfrac{1}{f'(x)} = \dfrac{1}{4} $ |
| Slope of $ f^{-1} $ directly | $ \dfrac{1}{4} $ ✓ |

The reciprocal-slope rule confirms: the slope of the inverse is exactly $ \tfrac{1}{4} $.

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x = sp.Symbol('x', real=True)

f = 4*x + 2
f_prime = sp.diff(f, x)

# Inverse: solve y = 4x+2 for x
y = sp.Symbol('y', real=True)
f_inv = sp.solve(f - y, x)[0]  # (y - 2) / 4
f_inv_prime = sp.diff(f_inv, y)

print(f"f(x)          = {f}")
print(f"f'(x)         = {f_prime}")
print(f"f_inv(y)      = {f_inv}")
print(f"(f_inv)'(y)   = {f_inv_prime}")
print(f"1 / f'(x)     = {1 / f_prime}  (reciprocal rule check)")

# --- plot ---
xs = np.linspace(-15, 15, 400)
ys = 4*xs + 2

ys_inv = np.linspace(-15, 15, 400)
xs_inv = (ys_inv - 2) / 4

fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(xs, ys,      label=r'$f(x) = 4x+2$')
ax.plot(ys_inv, xs_inv, label=r'$f^{-1}(y) = \frac{y-2}{4}$', linestyle='--')
ax.plot([-15, 15], [-15, 15], color='gray', linestyle=':', linewidth=1, label='y = x')

ax.set_xlim(-15, 15)
ax.set_ylim(-15, 15)
ax.axhline(0, color='black', linewidth=0.8)
ax.axvline(0, color='black', linewidth=0.8)
ax.grid(alpha=0.3)
ax.legend()
ax.set_title(r'$f(x)=4x+2$ and its inverse — slopes 4 and $\frac{1}{4}$')
ax.set_xlabel('input')
ax.set_ylabel('output')
plt.show()
```

## CLEP Practice Question

**Let $ f(x) = 4x + 2 $. If $ g = f^{-1} $, what is $ g'(10) $?**

(A) $ 4 $  
(B) $ 2 $  
(C) $ \dfrac{1}{4} $  
(D) $ \dfrac{1}{2} $  

<details>
<summary>Show solution</summary>

**Step 1 — Find the $ x $ value that maps to $ y = 10 $.**

Solve $ f(x) = 10 $:

$
4x + 2 = 10 \implies x = 2
$

So the point $ (2,\ 10) $ lies on $ f $, meaning $ (10,\ 2) $ lies on $ g = f^{-1} $.

**Step 2 — Apply the reciprocal-slope rule.**

$
g'(10) = \frac{1}{f'(2)}
$

Since $ f'(x) = 4 $ (constant), we have:

$
g'(10) = \frac{1}{4}
$

**Answer: (C) $ \dfrac{1}{4} $**
