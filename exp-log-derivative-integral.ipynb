## CLEP Summary: Derivatives and Integrals of Exponential & Logarithmic Functions  
#### Including $ b^x $, $ t e^t $, $ t^2 e^t $, and log combinations

---

## 1. Exponential Functions

### 1.1 Derivatives of $ e^t $ and $ e^{kt} $
$
\frac{d}{dt} e^t = e^t
$
$
\frac{d}{dt} e^{kt} = k e^{kt}
$

### 1.2 Integrals of $ e^t $ and $ e^{kt} $
$
\int e^t \, dt = e^t + C
$
$
\int e^{kt} \, dt = \frac{1}{k} e^{kt} + C
$

---

## 2. General Exponential Base $ b^x $

### 2.1 Derivative of $ b^x $
For any base $ b > 0, b \neq 1 $:

$
\frac{d}{dx} b^x = b^x \ln b
$

### 2.2 Integral of $ b^x $
$
\int b^x \, dx = \frac{b^x}{\ln b} + C
$

This is a **CLEP‑favorite identity**.

---

## 3. Logarithmic Functions

### 3.1 Derivatives
$
\frac{d}{dt} \ln t = \frac{1}{t}
$
$
\frac{d}{dt} \ln|f(t)| = \frac{f'(t)}{f(t)}
$

### 3.2 Integrals
$
\int \frac{1}{t} \, dt = \ln|t| + C
$
$
\int \frac{f'(t)}{f(t)} \, dt = \ln|f(t)| + C
$

---

## 4. Product Patterns with Exponentials  
These are high‑yield CLEP patterns requiring **integration by parts**.

---

## 4.1 Derivatives of $ t e^t $, $ t^2 e^t $, etc.

#### Product rule:
$
\frac{d}{dt}(t e^t) = (t+1)e^t
$

$
\frac{d}{dt}(t^2 e^t) = (t^2 + 2t)e^t
$

General pattern:
$
\frac{d}{dt}(t^n e^t) = e^t(t^n + n t^{n-1})
$

---

## 4.2 Integrals of $ t e^t $, $ t^2 e^t $, etc.

#### Example 1:
$
\int t e^t \, dt = (t - 1)e^t + C
$

#### Example 2:
$
\int t^2 e^t \, dt = (t^2 - 2t + 2)e^t + C
$

General pattern:
$
\int t^n e^t \, dt = e^t(t^n - n t^{n-1} + n(n-1)t^{n-2} - \cdots) + C
$

---

## 5. Logarithmic Combinations with Exponentials

### 5.1 Derivatives
$
\frac{d}{dt}(e^t \ln t)
= e^t\left(\ln t + \frac{1}{t}\right)
$

$
\frac{d}{dt}(t \ln t)
= \ln t + 1
$

$
\frac{d}{dt}(t^2 \ln t)
= 2t \ln t + t
$

---

### 5.2 Integrals involving logs

#### Classic pattern:
$
\int t \ln t \, dt
= \frac{t^2}{2}\ln t - \frac{t^2}{4} + C
$

## 7. CLEP Exam Nuances

- CLEP loves testing the identity:  
  $
  \frac{d}{dx} b^x = b^x \ln b
  $
- And the integral:  
  $
  \int b^x dx = \frac{b^x}{\ln b} + C
  $
- $ t e^t $ and $ t^2 e^t $ are **integration‑by‑parts classics**.  
- Logarithmic derivative rule is frequently tested:  
  $
  \frac{d}{dt}\ln|f(t)| = \frac{f'(t)}{f(t)}
  $
- CLEP often tests **pattern recognition**, not long computations.
