# Area Between Two Curves — CLEP Calculus Tutorial

This note explains how to compute the area between two curves using definite integrals.  
We cover the theory, two worked examples, and several CLEP‑style practice problems.

---

# 1. Core Idea

If one function is always above the other on an interval $[a,b]$, the area between them is:

$
A = \int_a^b \big( \text{top function} - \text{bottom function} \big)\, dx
$

This works because a definite integral represents **net area**, and subtracting the lower curve from the upper curve gives the **height** of each vertical slice.

---

# 2. When to Use x‑Integrals vs y‑Integrals

Most CLEP problems use vertical slices (integrating with respect to $x$):

- Use vertical slices when curves are functions of $x$: $y = f(x)$
- Use horizontal slices when curves are functions of $y$: $x = g(y)$

Horizontal slicing formula:

$
A = \int_{c}^{d} \big( \text{right function} - \text{left function} \big)\, dy
$

---

# 3. Example 1 — Vertical Slices

Find the area between:

$
y = x^2 \quad \text{and} \quad y = 4
$

Step 1: Find intersection points.

Solve:

$
x^2 = 4 \quad \Rightarrow \quad x = -2, 2
$

Step 2: Identify top and bottom.

- Top: $y = 4$
- Bottom: $y = x^2$

Step 3: Integrate.

$
A = \int_{-2}^{2} (4 - x^2)\, dx
$

Compute:

$
\int (4 - x^2)\, dx = 4x - \frac{x^3}{3}
$

Evaluate:

$
A = \left[4x - \frac{x^3}{3}\right]_{-2}^{2}
$

Plug in:

At $2$: $8 - \frac{8}{3} = \frac{16}{3}$

At $-2$: $-8 + \frac{8}{3} = -\frac{16}{3}$

Subtract:

$
A = \frac{16}{3} - \left(-\frac{16}{3}\right) = \frac{32}{3}
$

---

# 4. Example 2 — Horizontal Slices

Find the area between:

$
x = y^2 \quad \text{and} \quad x = 6 - y
$

Step 1: Find intersection points.

Solve:

$
y^2 = 6 - y
$

$
y^2 + y - 6 = 0
$

$
(y+3)(y-2)=0
$

So:

$
y = -3, \quad y = 2
$

Step 2: Identify right and left curves.

- Right: $x = 6 - y$
- Left: $x = y^2$

Step 3: Integrate with respect to $y$:

$
A = \int_{-3}^{2} \big( (6 - y) - y^2 \big)\, dy
$

Compute:

$
\int (6 - y - y^2)\, dy = 6y - \frac{y^2}{2} - \frac{y^3}{3}
$

Evaluate from $-3$ to $2$:

At $2$:  
$12 - 2 - \frac{8}{3} = 10 - \frac{8}{3} = \frac{22}{3}$

At $-3$:  
$-18 - \frac{9}{2} + 9 = -9 - \frac{9}{2} = -\frac{27}{2}$

Subtract:

$
A = \frac{22}{3} - \left(-\frac{27}{2}\right)
$

Common denominator $6$:

$
A = \frac{44}{6} + \frac{81}{6} = \frac{125}{6}
$

# 5. CLEP‑Style Practice Problems

y = x \quad \text{and} \quad y = x^2
$

on the interval where they intersect.

$

---

### Problem 4
Which integral correctly represents the area between  
$y = \sin x$ and $y = \cos x$ on $[0, \pi/4]$?

A.  $\int_0^{\pi/4} (\sin x - \cos x)\, dx$  
B.  $\int_0^{\pi/4} (\cos x - \sin x)\, dx$  
C.  $\int_0^{\pi/4} \sin x \cos x\, dx$  
D.  $\int_0^{\pi/4} |\sin x - \cos x|\, dx$

---

# 6. Solutions

### Solution 1

Intersection: solve $x = x^2$:

$
x(x-1)=0 \Rightarrow x=0,1
$

Top: $y=x$  
Bottom: $y=x^2$

$
A = \int_0^1 (x - x^2)\, dx = \left[\frac{x^2}{2} - \frac{x^3}{3}\right]_0^1 = \frac{1}{2} - \frac{1}{3} = \frac{1}{6}
$

---

### Solution 2

Solve $4 - x^2 = 2x$:

$
x^2 + 2x - 4 = 0
$

Roots:

$
x = -1 \pm \sqrt{5}
$

Top: $4 - x^2$  
Bottom: $2x$

$
A = \int_{-1-\sqrt{5}}^{-1+\sqrt{5}} \big(4 - x^2 - 2x\big)\, dx
$

---

### Solution 3

Intersection: solve $y = y^2$:

$
y(y-1)=0 \Rightarrow y=0,1
$

Right: $x=y$  
Left: $x=y^2$

$
A = \int_0^1 (y - y^2)\, dy = \left[\frac{y^2}{2} - \frac{y^3}{3}\right]_0^1 = \frac{1}{2} - \frac{1}{3} = \frac{1}{6}
$

---

### Solution 4

On $[0,\pi/4]$, $\cos x > \sin x$.

Correct answer: **B**
