### Optimization: Rectangle Under the Curve y = 8 - x^2
This is an application problem.
- Extra step of setting the problem up
- Remainder is the Math.
- Need to find an expression for the area.
- Once you have the expression, find the critical points and solve for the min/max as usual.
- Confusing part of this question is the domain. (e.g. Is the base of the rectangle $x$ or $2x$?)
- They want you to make it symmetric about the $y$-axis so use from $-x$, to $+x$ (or $2x$) 

Move the slider bar on the graph produced below and you can see then A is the largest.


```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Curve
def f(x):
    return 8 - x**2

# Interactive plot
def rectangle_plot(x):
    # Domain for the curve
    X = np.linspace(-3, 3, 400)
    Y = f(X)

    # Rectangle dimensions
    height = f(x)
    base_left = -x
    base_right = x

    # Clear figure
    plt.figure(figsize=(7,5))

    # Plot the curve
    plt.plot(X, Y, 'b', label='y = 8 - x^2')

    # Plot the rectangle
    rect_x = [base_left, base_right, base_right, base_left, base_left]
    rect_y = [0, 0, height, height, 0]
    plt.plot(rect_x, rect_y, 'r-', linewidth=2)
    plt.fill(rect_x, rect_y, color='red', alpha=0.3)

    # Labels and formatting
    plt.title(f"Rectangle for x = {x:.3f},  Area = {2*x*height:.3f}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.ylim(0, 10)
    plt.xlim(-3, 3)
    plt.grid(True)
    plt.legend()
    plt.show()

# Slider: x ranges from 0 to sqrt(8)
interact(rectangle_plot,
         x=FloatSlider(value=1.0, min=0.0, max=np.sqrt(8), step=0.05));
```



A rectangle is constructed with:
- its base on the x-axis,
- its top corners on the graph of $ y = 8 - x^2 $ for $ x \ge 0 $,
- and symmetric about the y-axis.

We will:
1. Express the area $ A(x) $ as a function of $ x $.
2. Find the critical points of $ A(x) $.
3. Determine the value of $ x $ that maximizes the area and find the maximum area.

---

## 0. Geometry and Domain

Let the right top corner of the rectangle be at $ (x, 8 - x^2) $ with $ x \ge 0 $.  
By symmetry, the left top corner is at $ (-x, 8 - x^2) $.

- Base of the rectangle: from $ -x $ to $ x $, so base length is $ 2x $.
- Height of the rectangle: $ y = 8 - x^2 $.

For the rectangle to exist, the height must be nonnegative:

$
8 - x^2 \ge 0 \quad \Rightarrow \quad x^2 \le 8 \quad \Rightarrow \quad 0 \le x \le \sqrt{8}.
$

So the **domain** of $ A(x) $ is:

$
0 \le x \le \sqrt{8}.
$

---

## 1. Area as a Function of x

Base:

$
\text{base} = 2x
$

Height:

$
\text{height} = 8 - x^2
$

Area:

$
A(x) = (\text{base})(\text{height}) = 2x(8 - x^2).
$

We can expand:

$
A(x) = 16x - 2x^3, \quad 0 \le x \le \sqrt{8}.
$

---

## 2. Critical Points of $ A(x) $

Differentiate:

$
A(x) = 16x - 2x^3
$

$
A'(x) = 16 - 6x^2.
$

Set derivative equal to zero:

$
16 - 6x^2 = 0
$

$
6x^2 = 16
$

$
x^2 = \frac{16}{6} = \frac{8}{3}
$

$
x = \sqrt{\frac{8}{3}}.
$

Since $ x \ge 0 $, the only critical point in the domain is:

$
x = \sqrt{\frac{8}{3}}.
$

---

## 3. Maximum Area

We check the area at:
- the endpoints $ x = 0 $ and $ x = \sqrt{8} $,
- the interior critical point $ x = \sqrt{\frac{8}{3}} $.

### At $ x = 0 $:

$
A(0) = 16 \cdot 0 - 2 \cdot 0^3 = 0.
$

### At $ x = \sqrt{8} $:

$
A(\sqrt{8}) = 2\sqrt{8}(8 - (\sqrt{8})^2) = 2\sqrt{8}(8 - 8) = 0.
$

### At $ x = \sqrt{\frac{8}{3}} $:

Use $ A(x) = 2x(8 - x^2) $:

$
A\!\left(\sqrt{\frac{8}{3}}\right)
= 2 \sqrt{\frac{8}{3}} \left(8 - \frac{8}{3}\right)
= 2 \sqrt{\frac{8}{3}} \cdot \frac{16}{3}
= \frac{32}{3} \sqrt{\frac{8}{3}}.
$

This is positive and larger than the endpoint values (which are 0), so it is the **maximum area**.

---

## Final Answers

1. **Area function:**
   $
   A(x) = 2x(8 - x^2) = 16x - 2x^3, \quad 0 \le x \le \sqrt{8}.
   $

2. **Critical point:**
   $
   x = \sqrt{\frac{8}{3}}.
   $

3. **Maximizing value and maximum area:**
   - The area is maximized at $ x = \sqrt{\frac{8}{3}} $.
   - The maximum area is
     $
     A_{\max} = \frac{32}{3} \sqrt{\frac{8}{3}}.
     $

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Curve
def f(x):
    return 8 - x**2

# Interactive plot
def rectangle_plot(x):
    # Domain for the curve
    X = np.linspace(-3, 3, 400)
    Y = f(X)

    # Rectangle dimensions
    height = f(x)
    base_left = -x
    base_right = x

    # Clear figure
    plt.figure(figsize=(7,5))

    # Plot the curve
    plt.plot(X, Y, 'b', label='y = 8 - x^2')

    # Plot the rectangle
    rect_x = [base_left, base_right, base_right, base_left, base_left]
    rect_y = [0, 0, height, height, 0]
    plt.plot(rect_x, rect_y, 'r-', linewidth=2)
    plt.fill(rect_x, rect_y, color='red', alpha=0.3)

    # Labels and formatting
    plt.title(f"Rectangle for x = {x:.3f},  Area = {2*x*height:.3f}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.ylim(0, 10)
    plt.xlim(-3, 3)
    plt.grid(True)
    plt.legend()
    plt.show()

# Slider: x ranges from 0 to sqrt(8)
interact(rectangle_plot,
         x=FloatSlider(value=1.0, min=0.0, max=np.sqrt(8), step=0.05));
```
