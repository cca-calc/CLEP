## **Inverse Functions & Tangent Line — CLEP Problem Explanation**

### **Given**
- $f$ and $g$ are inverses.  
- The line $y = 3x - 1$ is tangent to $g$ at $x = 2$.

### **Step 1: Extract information from the tangent line**
The slope of the tangent line is the slope of $g$ at that point:


$
g'(2) = 3.  
$
This is the step I missed!!


Find the point of tangency:


$
g(2) = 3(2) - 1 = 5.
$



### **Step 2: Use the inverse relationship**
Since $f$ and $g$ are inverses:


$
g(2) = 5 \quad \Longrightarrow \quad f(5) = 2.
$



### **Step 3: Use the derivative rule for inverses**


$
f'(g(x)) = \frac{1}{g'(x)}.
$
See below for derivation of this rule.


Plug in $x = 2$:


$
f'(5) = \frac{1}{g'(2)} = \frac{1}{3}.
$



### **Evaluate the statements**

#### **I. $f$ is decreasing at $x = 2$.**
We know nothing about $f'(2)$.  
**Not necessarily true.**

#### **II. $f'(5) = \frac{1}{3}$.**
We computed exactly this.  
**True.**

#### **III. $f'(2) = 3$.**
Derivative at $5$ is $\frac{1}{3}$, and we have no link to $f'(2)$.  
**False.**

### **Final Answer**
Only **Statement II** must be true.


# **Derivative Rule for Inverse Functions**

## **1. Core Identity**
Inverse functions satisfy:


$
f(f^{-1}(x)) = x.
$



Differentiate both sides with respect to $x$:

Left side (chain rule):


$
f'(f^{-1}(x)) \cdot (f^{-1})'(x)
$



Right side:


$
1
$



Solve for $(f^{-1})'(x)$:


$
(f^{-1})'(x) = \frac{1}{f'(f^{-1}(x))}.
$



---

## **2. Geometric Meaning**
Inverse functions **swap the axes**.

- A point $(a, b)$ on $f$ becomes $(b, a)$ on $f^{-1}$.
- Slopes are “rise over run.”
- Swapping axes swaps rise and run.

So if the slope of $f$ at $(a, b)$ is $m$, then the slope of $f^{-1}$ at $(b, a)$ is:


$
\frac{1}{m}.
$



---

## **3. How to Use the Rule**
If you know:
- a point on $g = f^{-1}$
- the slope of $g$ at that point

Then you immediately know:
- the corresponding point on $f$
- the slope of $f$ at that point

Example:
- If $g(2) = 5$ and $g'(2) = 3$, then  
  

$
  f(5) = 2, \quad f'(5) = \frac{1}{3}.
  $



---

## **4. When the Rule Fails**
You cannot use the inverse derivative rule if:
- $f$ is not one‑to‑one  
- $f'(x) = 0$ at the point  
- $f$ is not differentiable  

The inverse only has a slope where the original function has a **nonzero** slope.
