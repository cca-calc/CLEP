# Average Rate of Change

The **average rate of change** of a function $f(x)$ on an interval $[a,b]$ is defined as:

$
\frac{f(b) - f(a)}{b - a}
$

This quantity measures how much the function's output changes per unit change in the input over the interval.

---

## Interpretation

- You do not need to know calculus to calculate it.
- It is a definition, something made up.
- It is nothing more than the slope $\frac{\Delta y}{\Delta x}$ or $\frac{y_b - y_a}{x_b - x_a}$
- It is the **slope of the secant line** connecting the points $(a, f(a))$ and $(b, f(b))$.
- It describes the **overall change** in the function between $x=a$ and $x=b$.
- It does **not** depend on the behavior of the function inside the interval, only on the endpoints.

---

## Why it matters

The Mean Value Theorem states that if $f$ is continuous on $[a,b]$ and differentiable on $(a,b)$, then there exists some $c$ in $(a,b)$ such that:

$
f'(c) = \frac{f(b) - f(a)}{b - a}
$

This means the **instantaneous rate of change** (the derivative) matches the **average rate of change** somewhere inside the interval.

---

## Example

For $f(x) = x^2$ on $[1,4]$:

$
\frac{f(4) - f(1)}{4 - 1}
= \frac{16 - 1}{3}
= 5
$

So the function increases, on average, by **5 units of output per unit of input** over that interval.
