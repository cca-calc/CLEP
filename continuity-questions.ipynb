$

# Solutions: CLEP‑Style Removable Discontinuity Problems

---

## Problem 1 — Classic Rational Hole

$
f(x) = \frac{x^2 - 9}{x - 3}
$

1. **Where is $ f $ not defined?**

   Denominator zero when $ x - 3 = 0 \Rightarrow x = 3 $.  
   So $ f $ is not defined at $ x = 3 $.

2. **Does $ \lim_{x \to 3} f(x) $ exist?**

   Factor numerator:

   $
   \frac{x^2 - 9}{x - 3} = \frac{(x - 3)(x + 1)}{x - 3} = x + 1 \quad (x \neq 3)
   $

   Then:

   $
   \lim_{x \to 3} f(x) = \lim_{x \to 3} (x + 1) = 4
   $

   So the limit exists and equals 4.

3. **Redefine $ f(3) $ to make $ f $ continuous.**

   Set $ f(3) = 4 $.  
   This fills the hole and makes the function continuous at $ x = 3 $.

---

## Problem 2 — Piecewise Function With a Hole

$
f(x) = 
\begin{cases}
x^2 - 4, & x \neq 2 \\
k, & x = 2
\end{cases}
$

1. **Does $ \lim_{x \to 2} f(x) $ exist?**

   For $ x \neq 2 $, $ f(x) = x^2 - 4 $. Then:

   $
   \lim_{x \to 2} f(x) = \lim_{x \to 2} (x^2 - 4) = 4 - 4 = 0
   $

   So the limit exists and equals 0.

2. **Find $ k $ for continuity at $ x = 2 $.**

   Continuity requires:

   $
   f(2) = \lim_{x \to 2} f(x) = 0
   $

   So $ k = 0 $.

3. **Type of discontinuity (before choosing $ k $)?**

   The limit exists but $ f(2) $ may not equal the limit (or may be undefined).  
   This is a **removable discontinuity**.

---

## Problem 3 — Hidden Hole in a Simplified Expression

$
f(x) = \frac{(x-5)(x+1)}{x-5}
$

1. **Identify the discontinuity.**

   Denominator zero when $ x - 5 = 0 \Rightarrow x = 5 $.  
   So $ f $ is not defined at $ x = 5 $: a hole.

2. **Evaluate $ \lim_{x \to 5} f(x) $.**

   For $ x \neq 5 $:

   $
   f(x) = x + 1
   $

   So:

   $
   \lim_{x \to 5} f(x) = \lim_{x \to 5} (x + 1) = 6
   $

3. **Why does the limit exist?**

   Because near $ x = 5 $, $ f(x) $ behaves exactly like $ x + 1 $, which approaches 6.  
   The missing value at $ x = 5 $ does not affect the approaching behavior.

## Problem 4 — Solve for Continuity in a Piecewise Function

$
f(x) =
\begin{cases}
3x + 1, & x < 2 \\
ax^2 - 2, & x \ge 2
\end{cases}
$

Continuity at $ x = 2 $ requires:

$
\lim_{x \to 2^-} f(x) = \lim_{x \to 2^+} f(x) = f(2)
$

Left side:

$
\lim_{x \to 2^-} f(x) = 3(2) + 1 = 7
$

Right side and value at 2:

$
f(2) = a(2^2) - 2 = 4a - 2
$

Set equal:

$
4a - 2 = 7 \Rightarrow 4a = 9 \Rightarrow a = \frac{9}{4}
$

## Problem 5 — Limit Exists but Function Value Is Wrong

$
f(x) =
\begin{cases}
\frac{x^2 - 16}{x - 4}, & x \neq 4 \\
10, & x = 4
\end{cases}
$

1. **Compute $ \lim_{x \to 4} f(x) $.**

   For $ x \neq 4 $:

   $
   \frac{x^2 - 16}{x - 4} = \frac{(x - 4)(x + 1)}{x - 4} = x + 4
   $

   So:

   $
   \lim_{x \to 4} f(x) = \lim_{x \to 4} (x + 4) = 8
   $

2. **Is $ f $ continuous at $ x = 4 $?**

   $ f(4) = 10 $, but the limit is 8.  
   Since $ f(4) \neq \lim_{x \to 4} f(x) $, the function is **not continuous** at 4.

3. **Type of discontinuity?**

   The limit exists but the function value is wrong → **removable discontinuity**.

4. **Redefine $ f(4) $ to fix it.**

   Set $ f(4) = 8 $.

---

## Problem 6 — Determine Continuity From a Graph (Conceptual)

Graph: smooth curve with an open circle at $ (1, 3) $, and no value defined at $ x = 1 $.

1. **Does $ \lim_{x \to 1} f(x) $ exist?**

   Yes. The graph approaches height 3 from both sides, so:

   $
   \lim_{x \to 1} f(x) = 3
   $

2. **Is the function continuous at $ x = 1 $?**

   No. $ f(1) $ is not defined, so continuity fails.

3. **Type of discontinuity?**

   **Removable discontinuity** (a hole).

4. **How to make it continuous?**

   Define $ f(1) = 3 $.

---

## Problem 7 — One‑Sided Limit Check

$
f(x) =
\begin{cases}
2x - 1, & x < 1 \\
x + 2, & x > 1
\end{cases}
$

1. **Does $ \lim_{x \to 1} f(x) $ exist?**

   Left‑hand limit:

   $
   \lim_{x \to 1^-} f(x) = 2(1) - 1 = 1
   $

   Right‑hand limit:

   $
   \lim_{x \to 1^+} f(x) = 1 + 2 = 3
   $

   Since $ 1 \neq 3 $, the limit does **not** exist.

2. **Is there a removable discontinuity at $ x = 1 $?**

   No. A removable discontinuity requires the left and right limits to agree.  
   Here they do not, so this is a **jump discontinuity**, not removable.

3. **Value of $ f(1) $ to make it continuous?**

   Impossible: no single value can make the left and right limits match.  
   So no choice of $ f(1) $ makes the function continuous.

---

## Problem 8 — Algebraic Limit Leading to a Hole

$
f(x) = \frac{\sin(x)}{x}
$

1. **Where is $ f $ not defined?**

   Denominator zero at $ x = 0 $. So $ f $ is not defined at 0.

2. **Does $ \lim_{x \to 0} f(x) $ exist?**

   Yes, a standard limit:

   $
   \lim_{x \to 0} \frac{\sin(x)}{x} = 1
   $

3. **Type of discontinuity at $ x = 0 $?**

   The limit exists but the function is undefined at 0 → **removable discontinuity**.

4. **Redefine $ f(0) $ to make it continuous.**

   Set $ f(0) = 1 $.

---

## Problem 9 — Parameter Matching for Continuity

$
f(x) =
\begin{cases}
x + 4, & x \le 1 \\
bx^2 - 1, & x > 1
\end{cases}
$

Continuity at $ x = 1 $ requires:

$
\lim_{x \to 1^-} f(x) = \lim_{x \to 1^+} f(x) = f(1)
$

Left side and value at 1:

$
f(1) = 1 + 4 = 5
$

Right side:

$
\lim_{x \to 1^+} f(x) = b(1^2) - 1 = b - 1
$

Set equal:

$
b - 1 = 5 \Rightarrow b = 6
$

## Problem 10 — CLEP‑Style Multiple Choice

$
f(x) = \frac{x^2 - 25}{x - 5}
$

Factor:

$
\frac{x^2 - 25}{x - 5} = \frac{(x - 5)(x + 5)}{x - 5} = x + 5 \quad (x \neq 5)
$

- $ f $ is not defined at $ x = 5 $ (hole).
- The limit:

  $
  \lim_{x \to 5} f(x) = \lim_{x \to 5} (x + 5) = 10
  $

- So the limit exists and equals 10, but the function is not continuous at $ x = 5 $.

Correct choice:  
**B.** The limit exists and equals 10, but the function is not continuous at $ x = 5 $.
