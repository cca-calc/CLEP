# Using cos(x) for shape of curve

We will explore the following for $ f(x) = \cos(x) $:

- Critical points
- Local maxima and minima
- Inflection points
- Concavity intervals
- Visual confirmation with plots

On the interval $[ -2\pi, 2\pi ]$.


```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

sp.init_printing()

x = sp.symbols('x', real=True)
f = sp.cos(x)
f
```

## Plot of $ \cos(x) $ on $[ -2\pi, 2\pi ]$

This gives a visual sense of where maxima, minima, and inflection points occur.


```python
x_vals = np.linspace(-2*np.pi, 2*np.pi, 400)
y_vals = np.cos(x_vals)

plt.figure(figsize=(8, 4))
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.plot(x_vals, y_vals, label=r'$\cos(x)$')
plt.title(r'$f(x) = \cos(x)$ on $[-2\pi, 2\pi]$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

## First derivative and critical points

For $ f(x) = \cos(x) $:



$
f'(x) = -\sin(x).
$



**Critical points** occur where $ f'(x) = 0 $ or $ f'(x) $ is undefined.  
Here, $ -\sin(x) = 0 \Rightarrow \sin(x) = 0 \Rightarrow x = k\pi $ for any integer $ k $.

On $[ -2\pi, 2\pi ]$, the critical points are:



$
x \in \{ -2\pi, -\pi, 0, \pi, 2\pi \}.
$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

plt.figure(figsize=(10, 5))
plt.plot(x, np.cos(x), label='cos(x)', linewidth=2)
plt.plot(x, -np.sin(x), label='-sin(x)', linewidth=2 , linestyle='--')
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.xlim(-2 * np.pi, 2 * np.pi)
plt.xticks(
    [-2 * np.pi, -np.pi, 0, np.pi, 2 * np.pi],
    [r'$-2\pi$', r'$-\pi$', '0', r'$\pi$', r'$2\pi$']
)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.title(r'cos(x), -sin(x) on $[-2\pi, 2\pi]$')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## Second derivative and classification of extrema

For $ f(x) = \cos(x) $:



$
f''(x) = -\cos(x).
$



We use the **second derivative test**:

- If $ f''(x_c) < 0 $, then $ x_c $ is a local maximum.
- If $ f''(x_c) > 0 $, then $ x_c $ is a local minimum.
- If $ f''(x_c) = 0 $, the test is inconclusive.

We evaluate $ f''(x) $ at the critical points $ x = k\pi $.


$
f''(x) = -\cos(x).
$



- If $ f''(x) > 0 \Rightarrow -\cos(x) > 0 $: **concave up**.
- If $ f''(x) < 0 \Rightarrow -\cos(x) < 0 $: **concave down**.

**Inflection points** occur where $ f''(x) = 0 $ and the concavity changes sign.



$
f''(x) = -\cos(x) = 0 \Rightarrow \cos(x) = 0 \Rightarrow x = \frac{\pi}{2} + k\pi.
$



On $[ -2\pi, 2\pi ]$, these are:



$
x \in \left\{ -\frac{3\pi}{2}, -\frac{\pi}{2}, \frac{\pi}{2}, \frac{3\pi}{2} \right\}.
$




### Concavity intervals on $[ -2\pi, 2\pi ]$

We know $ \cos(x) $ changes sign at the inflection points.  
Using the sign of $ \cos(x) $:

- **Concave down** where $ \cos(x) > 0 $.
- **Concave up** where $ \cos(x) < 0 $.

On $[ -2\pi, 2\pi ]$, the intervals are:

- Concave down:  
  

$
  (-2\pi, -\tfrac{3\pi}{2}),\quad (-\tfrac{\pi}{2}, \tfrac{\pi}{2}),\quad (\tfrac{3\pi}{2}, 2\pi)
  $


- Concave up:  
  

$
  (-\tfrac{3\pi}{2}, -\tfrac{\pi}{2}),\quad (\tfrac{\pi}{2}, \tfrac{3\pi}{2})
  $




## Visualizing critical points and inflection points

We now overlay:

- Critical points (local maxima/minima) in red.
- Inflection points in green.


```python
x_vals = np.linspace(-2*np.pi, 2*np.pi, 400)
y_vals = np.cos(x_vals)

plt.figure(figsize=(8, 4))
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.plot(x_vals, y_vals, label=r'$\cos(x)$')

# Critical points
# Critical points in [-2π, 2π]
crit_x = [-2*np.pi, -np.pi, 0, np.pi, 2*np.pi]

# Inflection points in [-2π, 2π]
inflection_points = [-3*np.pi/2, -np.pi/2, np.pi/2, 3*np.pi/2]
crit_y = [np.cos(cx) for cx in crit_x]
plt.scatter(crit_x, crit_y, color='red', zorder=5, label='critical points')

# Inflection points
infl_x = [float(ip) for ip in inflection_points]
infl_y = [np.cos(ix) for ix in infl_x]
plt.scatter(infl_x, infl_y, color='green', zorder=5, label='inflection points')

plt.title(r'Critical and inflection points of $f(x) = \cos(x)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

## Summary for $ f(x) = \cos(x) $

- **Function:** $ f(x) = \cos(x) $
- **First derivative:** $ f'(x) = -\sin(x) $
- **Second derivative:** $ f''(x) = -\cos(x) $

- **Critical points:** $ x = k\pi $.  
  On $[ -2\pi, 2\pi ]$: $ -2\pi, -\pi, 0, \pi, 2\pi $.

- **Local maxima:** at $ x = 2k\pi $ (even multiples of $ \pi $), where $ f(x) = 1 $.  
- **Local minima:** at $ x = (2k+1)\pi $ (odd multiples of $ \pi $), where $ f(x) = -1 $.

- **Inflection points:** $ x = \frac{\pi}{2} + k\pi $.  
  On $[ -2\pi, 2\pi ]$: $ -\tfrac{3\pi}{2}, -\tfrac{\pi}{2}, \tfrac{\pi}{2}, \tfrac{3\pi}{2} $.

- **Concavity:**
  - Concave down where $ \cos(x) > 0 $.
  - Concave up where $ \cos(x) < 0 $.


```python
import numpy as np
import matplotlib.pyplot as plt

x_plot = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

plt.figure(figsize=(10, 5))
plt.plot(x_plot, np.cos(x_plot), label='cos(x)', linewidth=2, color="green")
plt.plot(x_plot, -np.sin(x_plot), label='-sin(x)', linewidth=2 , linestyle='--')
plt.plot(x_plot, -np.cos(x_plot), label='-cos(x)', linewidth=2, color='red')
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.xlim(-2 * np.pi, 2 * np.pi)
plt.xticks(
    [-2 * np.pi, -np.pi, 0, np.pi, 2 * np.pi],
    [r'$-2\pi$', r'$-\pi$', '0', r'$\pi$', r'$2\pi$']
)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.title(r'cos(x), -sin(x), and -cos(x) on $[-2\pi, 2\pi]$')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
