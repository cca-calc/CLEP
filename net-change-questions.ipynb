# Net Change Theorem — Questions ChatGPT 5.1

These problems focus on the Net Change Theorem:

$
\text{Net Change} = \int_a^b \text{(rate)}\,dt
$

Each problem uses a rate function and asks for the total accumulated change.




## Problems 1–5

### **1. Velocity → Displacement**
A particle moves with velocity  
$
v(t)=t^2-4t+5.
$  
Find the net displacement from $t=0$ to $t=6$.

---

### **2. Water Flow Into a Tank**
Water flows into a tank at rate  
$
r(t)=12e^{-0.2t}.
$  
How much water enters the tank during the first 10 minutes?

---

### **3. Cooling Rate**
A metal rod cools at rate  
$
T'(t)=-7e^{-0.5t}.
$  
Find the total temperature drop from $t=0$ to $t=8$.

---

### **4. Population Growth**
A population grows at rate  
$
P'(t)=300-20t.
$  
Find the net population change from $t=0$ to $t=12$.

---

### **5. Car Acceleration → Change in Velocity**
A car’s acceleration is  
$
a(t)=4\sin(t)+2.
$  
Find the change in velocity from $t=0$ to $t=\pi$.

## Problems 6–10

### **6. Chemical Reaction Rate**
A reaction produces a substance at rate  
$
R(t)=\frac{20}{1+t^2}.
$  
How much product is formed from $t=0$ to $t=5$?

---

### **7. River Flow**
The flow rate of a river past a sensor is  
$
F(t)=50+10\cos\left(\frac{\pi t}{12}\right).
$  
Find the total volume of water passing the sensor from $t=0$ to $t=24$.

---

### **8. Heating Power → Energy**
A heater outputs power  
$
P(t)=1000+200\sin(0.5t).
$  
Find the total energy delivered from $t=0$ to $t=10$.

---

### **9. Rate Graph (Piecewise Linear)**
A rate function $f(t)$ is defined by the graph:

- From $t=0$ to $t=3$: line from $(0,0)$ to $(3,6)$  
- From $t=3$ to $t=5$: constant at $6$  
- From $t=5$ to $t=8$: line from $(5,6)$ to $(8,0)$

Compute the net change from $t=0$ to $t=8$.

---

### **10. Net Change with Negative Rates**
A tank leaks water at rate  
$
L(t)=5+3\sin t.
$  
How much water is lost from $t=0$ to $t=4\pi$?

## Problems 11–15

### **11. Electric Current → Charge**
An electric current flows according to  
$
I(t)=2t+1.
$  
Find the total charge transferred from $t=0$ to $t=10$.

---

### **12. Speed of a Conveyor Belt**
Boxes move along a conveyor with speed  
$
v(t)=8-0.4t.
$  
Find the distance traveled from $t=0$ to $t=15$.

---

### **13. Rate of Sand Falling**
Sand falls onto a surface at rate  
$
S(t)=10e^{-t/3}+2.
$  
Find the total sand deposited from $t=0$ to $t=12$.

---

### **14. Rate of Medication in Bloodstream**
A medication enters the bloodstream at rate  
$
M(t)=\frac{50t}{(1+t^2)^2}.
$  
Find the total amount absorbed from $t=0$ to $t=6$.

---

### **15. Abstract Function via Derivative**
You are told that  
$
F'(x)=\cos(x)-\frac{x}{4}.
$  
Find the net change in $F$ from $x=0$ to $x=4$.

# Solutions

### **1.**  
$
\int_0^6 (t^2 - 4t + 5)\,dt = 42.
$

### **2.**  
$
\int_0^{10} 12e^{-0.2t}\,dt = 60(1 - e^{-2}).
$

### **3.**  
$
\int_0^8 -7e^{-0.5t}\,dt = -14(1 - e^{-4}).
$

### **4.**  
$
\int_0^{12} (300 - 20t)\,dt = 1440.
$

### **5.**  
$
\int_0^\pi (4\sin t + 2)\,dt = 2\pi + 8.
$

### **6.**  
$
\int_0^5 \frac{20}{1+t^2}\,dt = 20\arctan(5).
$

### **7.**  
$
\int_0^{24} \left(50 + 10\cos\left(\frac{\pi t}{12}\right)\right) dt = 1200.
$

### **8.**  
$
\int_0^{10} (1000 + 200\sin(0.5t))\,dt = 10000 + 800(1 - \cos 5).
$

### **9.**  
Triangle + rectangle + triangle  
$
9 + 12 + 9 = 30.
$

### **10.**  
$
\int_0^{4\pi} (5 + 3\sin t)\,dt = 20\pi.
$

### **11.**  
$
\int_0^{10} (2t + 1)\,dt = 120.
$

### **12.**  
$
\int_0^{15} (8 - 0.4t)\,dt = 90.
$

### **13.**  
$
\int_0^{12} (10e^{-t/3} + 2)\,dt = 30(1 - e^{-4}) + 24.
$

### **14.**  
$
\int_0^6 \frac{50t}{(1+t^2)^2}\,dt = 25\left(1 - \frac{1}{37}\right).
$

### **15.**  
$
\int_0^4 \left(\cos x - \frac{x}{4}\right)\,dx = \sin 4 - 2.
$
