# Solving the Differential Equation  dy/dx = k y

We want to solve the differential equation

$
\frac{dy}{dx} = k y
$

This equation appears in population growth, radioactive decay, compound interest, and many CLEPâ€‘level problems.

---

## Deriving the Solution

Start with:

$
\frac{dy}{dx} = k y
$

Separate variables:

$
\frac{1}{y} \, dy = k \, dx
$

Integrate both sides:

$
\int \frac{1}{y} \, dy = \int k \, dx
$

This gives:

$
\ln |y| = kx + C
$

Exponentiate:

$
|y| = e^{kx + C} = e^C e^{kx}
$

Let $ e^C = A $. Allowing the constant to be positive or negative:

$
y = C e^{kx}
$

This is the **general solution**.

---

## Why Exponential Functions Work

The equation says the rate of change of $y$ is proportional to $y$:

$
\frac{dy}{dx} = k y
$

Exponential functions uniquely satisfy:

$
\frac{d}{dx}(e^{kx}) = k e^{kx}
$

So exponentials are the only functions whose derivative is a constant multiple of themselves.

---

## Example 1

Solve:

$
\frac{dy}{dx} = 5y
$

General solution:

$
y = C e^{5x}
$

If $y(0)=3$:

$
3 = C e^0 = C
$

Final:

$
y = 3 e^{5x}
$

---

## Example 2

Solve:

$
\frac{dy}{dx} = -2y, \quad y(1)=10
$

General solution:

$
y = C e^{-2x}
$

Apply the initial condition:

$
10 = C e^{-2}
$

So:

$
C = 10 e^{2}
$

Final:

$
y = 10 e^{2} e^{-2x}
$

1. Solve  
   $

   $

2. Solve  
   $
   \frac{dy}{dx} = -3y, \quad y(2)=5
   $

3. Which function satisfies $dy/dx = ky$?  
   A. $kx^2$  
   B. $e^{kx}$  
   C. $\ln(kx)$  
   D. $\sin(kx)$

4. Verify that $y = C e^{kx}$ satisfies $dy/dx = ky$.

---

## Solutions

**1.**  
$
y = 4 e^{7x}
$

**2.**  
$
y = 5 e^{6} e^{-3x}
$

**3.**  
Correct answer: **B**

**4.**  
Differentiate:  
$
\frac{dy}{dx} = C k e^{kx} = k(C e^{kx}) = k y
$
Verified.
