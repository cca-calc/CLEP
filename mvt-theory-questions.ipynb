```python
# Graph for Question 12
x = np.linspace(0, 6, 400)
y = np.piecewise(x,
                 [x < 3, x >= 3],
                 [lambda t: 2*t, lambda t: 6])

plt.figure(figsize=(5,4))
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.plot(x[x<3], y[x<3], label="f(x) on [0,3)")
plt.plot(x[x>=3], y[x>=3], label="f(x) on [3,6]")
plt.scatter([3],[6], color='black')
plt.title("Figure 2: Graph of a piecewise function on [0,6]")
plt.xlim(0,6)
plt.ylim(0,13)
plt.grid(True)
plt.legend()
plt.show()
```
