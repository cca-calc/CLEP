# The Net Change Theorem

The Net Change Theorem states that if a quantity has a rate of change given by a function $ f(x) $, then the total change in the quantity from $ x = a $ to $ x = b $ is:

$
\int_a^b f(x)\,dx = F(b) - F(a)
$

where $ F $ is any antiderivative of $ f $.

**Notes:**
- Integrating a **rate** gives the **total accumulated change**. 
- The rate is the derivate with respect to the integration variable (e.g. velocity)
- Signed area above the axis increases the quantity.
- Signed area below the axis decreases it.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
from scipy.integrate import quad

# Piecewise rate function
def f(x):
    return np.piecewise(
        x,
        [x < 2, (x >= 2) & (x < 5), x >= 5],
        [
            lambda x: x,      # increasing
            lambda x: 2,      # constant
            lambda x: -x + 7  # decreasing
        ]
    )

x = np.linspace(0, 7, 400)
a, b = 0, 7

plt.figure(figsize=(10, 4))
plt.plot(x, f(x), label='f(x) = rate')
plt.axhline(0, color='black', linewidth=0.8)
plt.title('Rate Function: Piecewise Linear')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()

net_change, _ = quad(f, a, b)
print('Net change:', net_change)
```

```python
plt.figure(figsize=(10, 4))
plt.plot(x, f(x), label='f(x)')
plt.fill_between(x, f(x), alpha=0.3)
plt.axhline(0, color='black', linewidth=0.8)
plt.title('Net Change = Signed Area Under f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

# Motion Interpretation

If $ v(t) $ is velocity, then the displacement from $ t = a $ to $ t = b $ is:

$
\int_a^b v(t)\,dt
$

This is the most intuitive example of the Net Change Theorem:

- Velocity is the **rate of change of position**.
- Integrating velocity gives **total change in position**.

```python
v = lambda t: 3 + 2*np.sin(t / 2) 
t = np.linspace(0, 10, 400)

plt.figure(figsize=(10, 4))
plt.plot(t, v(t), label='v(t) = velocity')
plt.fill_between(t, v(t), alpha=0.3)
plt.axhline(0, color='black', linewidth=0.8)
plt.title('Displacement = Area Under Velocity Curve')
plt.xlabel('t (seconds)')
plt.ylabel('v(t)')
plt.grid(True)
plt.legend()
plt.show()

a = 0
b = 10

net_change, _ = quad( v, a, b)

print('Net change in velocity:', net_change)
net_change
absolute_change, _ = quad(lambda t: np.abs(v(t)), a, b)
print('Absolute change in velocity:', absolute_change)
```

```python
v = lambda t: 3 + 2*np.sin(t / 2) - 2
t = np.linspace(0, 10, 400)

plt.figure(figsize=(10, 4))
plt.plot(t, v(t), label='v(t) = velocity')
plt.fill_between(t, v(t), alpha=0.3)
plt.axhline(0, color='black', linewidth=0.8)
plt.title('Displacement = Area Under Velocity Curve')
plt.xlabel('t (seconds)')
plt.ylabel('v(t)')
plt.grid(True)
plt.legend()
plt.show()

a = 0
b = 10

net_change, _ = quad( v, a, b)

print('Net change in velocity:', net_change)
net_change
absolute_change, _ = quad(lambda t: np.abs(v(t)), a, b)
print('Absolute change in velocity:', absolute_change)
```

# Connection to Antiderivatives

If $ F'(x) = f(x) $, then:

$
\int_a^b f(x)\,dx = F(b) - F(a)
$

This is the Fundamental Theorem of Calculus in action.

The Net Change Theorem is simply the applied interpretation:

- **Integrals measure accumulated change.**

# Practice Problems

1. A rate function is given by $ f(x) = 4 - x $. Find the net change from $ x = 0 $ to $ x = 4 $.

2. A tank fills at rate $ r(t) = 10e^{-t} $. How much water enters from $ t = 0 $ to $ t = 5 $?

3. The velocity of a particle is $ v(t) = t^2 - 3t + 2 $. Find the displacement from $ t = 0 $ to $ t = 3 $.

4. A population grows at rate $ P'(t) = 500 - 20t $. Find the total growth from $ t = 0 $ to $ t = 20 $.

5. The graph of a rate function dips below the axis between $ x = 3 $ and $ x = 6 $. Explain what this means in terms of net change.

# Solutions

1. $\int_0^4 (4 - x)\,dx = 8.$

2. $\int_0^5 10e^{-t}\,dt = 10(1 - e^{-5}).$

3. $\int_0^3 (t^2 - 3t + 2)\,dt = 1.5.$

4. $\int_0^{20} (500 - 20t)\,dt = 3000.$

5. Negative area means the quantity is **decreasing** during that interval (the rate is negative, so the net change over that part is a decrease).
