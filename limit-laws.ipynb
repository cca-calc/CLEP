# Limit Laws (1–9)

## 1. Constant Law
$$\lim_{x \to a} c = c$$

## 2. Identity Law
$$\lim_{x \to a} x = a$$

## 3. Sum Law
$$\lim_{x \to a} \left[f(x) + g(x)\right] 
= \lim_{x \to a} f(x) + \lim_{x \to a} g(x)$$

## 4. Difference Law
$$\lim_{x \to a} \left[f(x) - g(x)\right] 
= \lim_{x \to a} f(x) - \lim_{x \to a} g(x)$$

## 5. Constant Multiple Law
$$\lim_{x \to a} \left[c\,f(x)\right] 
= c \cdot \lim_{x \to a} f(x)$$

## 6. Product Law
$$\lim_{x \to a} \left[f(x)g(x)\right] 
= \left(\lim_{x \to a} f(x)\right)\left(\lim_{x \to a} g(x)\right)$$

## 7. Quotient Law
$$\lim_{x \to a} \frac{f(x)}{g(x)} 
= \frac{\lim_{x \to a} f(x)}{\lim_{x \to a} g(x)} \quad \text{if } \lim_{x \to a} g(x) \neq 0$$

## 8. Power Law
$$\lim_{x \to a} \left[f(x)\right]^n 
= \left(\lim_{x \to a} f(x)\right)^n$$

## 9. Root Law
$$\lim_{x \to a} \sqrt[n]{f(x)} 
= \sqrt[n]{\lim_{x \to a} f(x)}$$
