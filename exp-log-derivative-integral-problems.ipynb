#### Problem 10 ‚Äî Multiple Choice (Pattern Recognition)

Which of the following is an antiderivative of $ 2^x $?

A. $ 2^x + C $  
B. $ x 2^x + C $  
C. $ \dfrac{2^x}{\ln 2} + C $  
D. $ (\ln 2) 2^x + C $

### Part B ‚Äî Solutions

---

#### Solution 1 ‚Äî Derivative of $ 7 b^x $

$
\frac{d}{dx}(7 b^x)
= 7 \cdot \frac{d}{dx}(b^x)
= 7 b^x \ln b
$

---

#### Solution 2 ‚Äî Integral of $ 3 \cdot 2^x $

$
\int 3 \cdot 2^x \, dx
= 3 \int 2^x \, dx
= 3 \cdot \frac{2^x}{\ln 2} + C
= \frac{3 \cdot 2^x}{\ln 2} + C
$

---

#### Solution 3 ‚Äî Exponential with Linear Argument

**Derivative:**

$
\frac{d}{dt}(5 e^{4t})
= 5 \cdot 4 e^{4t}
= 20 e^{4t}
$

**Integral:**

$
\int e^{4t} \, dt
= \frac{1}{4} e^{4t} + C
$

---

#### Solution 4 ‚Äî Logarithmic Derivative

$
\frac{d}{dx} \ln(5x^2 + 3)
= \frac{1}{5x^2 + 3} \cdot \frac{d}{dx}(5x^2 + 3)
= \frac{10x}{5x^2 + 3}
$

---

#### Solution 5 ‚Äî Logarithmic Integral Pattern

$
\int \frac{10x}{5x^2 + 7} \, dx
$

Note:
$
\frac{d}{dx}(5x^2 + 7) = 10x
$

So:
$
\int \frac{10x}{5x^2 + 7} \, dx
= \ln|5x^2 + 7| + C
$

---

#### Solution 6 ‚Äî Product $ t e^t $

**1. Derivative**

$
\frac{d}{dt}(t e^t)
= t e^t + e^t
= (t + 1)e^t
$

**2. Integral**

Use integration by parts: $ u = t,\; dv = e^t dt \Rightarrow du = dt,\; v = e^t $.

$
\int t e^t \, dt
= t e^t - \int e^t \, dt
= t e^t - e^t + C
= (t - 1)e^t + C
$

---

#### Solution 7 ‚Äî Product $ t^2 e^t $

First integration by parts: $ u = t^2,\; dv = e^t dt \Rightarrow du = 2t dt,\; v = e^t $.

$
\int t^2 e^t \, dt
= t^2 e^t - \int 2t e^t \, dt
$

Now integrate $ \int 2t e^t dt $ by parts again: $ u = 2t,\; dv = e^t dt \Rightarrow du = 2 dt,\; v = e^t $.

$
\int 2t e^t \, dt
= 2t e^t - \int 2 e^t \, dt
= 2t e^t - 2e^t
$

Substitute back:

$
\int t^2 e^t \, dt
= t^2 e^t - (2t e^t - 2e^t)
= t^2 e^t - 2t e^t + 2e^t
= (t^2 - 2t + 2)e^t + C
$

---

#### Solution 8 ‚Äî Log with Polynomial

$
\int t \ln t \, dt,\quad t > 0
$

Integration by parts: $ u = \ln t,\; dv = t dt \Rightarrow du = \frac{1}{t} dt,\; v = \frac{t^2}{2} $.

$
\int t \ln t \, dt
= \frac{t^2}{2}\ln t - \int \frac{t^2}{2} \cdot \frac{1}{t} dt
= \frac{t^2}{2}\ln t - \frac{1}{2}\int t \, dt
$
$
= \frac{t^2}{2}\ln t - \frac{1}{2} \cdot \frac{t^2}{2}
= \frac{t^2}{2}\ln t - \frac{t^2}{4} + C
$

---

#### Solution 9 ‚Äî Mixed Exponential and Log

$
\frac{d}{dt}(e^t \ln t),\quad t > 0
$

Product rule:

$
\frac{d}{dt}(e^t \ln t)
= e^t \ln t + e^t \cdot \frac{1}{t}
= e^t\left(\ln t + \frac{1}{t}\right)
$

---

#### Solution 10 ‚Äî Multiple Choice

We know:
$
\int 2^x \, dx = \frac{2^x}{\ln 2} + C
$

So the correct answer is:

**C. $ \dfrac{2^x}{\ln 2} + C $**
