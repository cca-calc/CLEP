### CLEP Calculus Summary: Functions, Curve Shapes, Critical Points, Extrema, and Derivative Tests

This guide summarizes the core concepts the CLEP exam expects you to know about how functions behave, how derivatives shape curves, and how to identify maxima, minima, and inflection points.

---

### 1. Functions and Their Behavior

### 1.1 Domain and Range
- **Domain**: all allowable inputs.
- **Range**: all possible outputs.
- CLEP often asks: “Where is the function continuous?” → Identify domain restrictions.

### 1.2 Common Function Families
- **Polynomials**: smooth, continuous, differentiable everywhere.
- **Rational functions**: discontinuous where denominator = 0.
- **Exponentials/logarithms**: continuous on natural domains.
- **Trig functions**: periodic behavior; discontinuities in tan, sec, csc, cot.
- **Piecewise functions**: continuity must be checked at boundary points.

---

### 2. The Derivative as a Tool for Understanding Shape

### 2.1 First Derivative: Slope and Increasing/Decreasing
- $ f'(x) > 0 $: function increasing.
- $ f'(x) < 0 $: function decreasing.
- CLEP frequently tests sign charts and interpreting derivative graphs.

### 2.2 Second Derivative: Concavity
- $ f''(x) > 0 $: concave up (cup‑shaped).
- $ f''(x) < 0 $: concave down (cap‑shaped).

### 2.3 Inflection Points
- Occur where concavity changes sign.
- CLEP nuance:  
  - $ f''(x) = 0 $ is **not enough**; concavity must actually change.

---

### 3. Critical Points

### 3.1 Definition
A **critical point** occurs where:
- $ f'(x) = 0 $ (horizontal tangent), or
- $ f'(x) $ does not exist, but $ f(x) $ is defined.

### 3.2 Why They Matter
Critical points are **candidates** for:
- Local maxima
- Local minima
- Inflection points (sometimes)

CLEP nuance:  
Not every critical point is an extremum (e.g., $ f(x) = x^3 $ at $ x = 0 $).

---

### 4. Local Maxima and Minima

### 4.1 First Derivative Test
Check the sign of $ f'(x) $ around a critical point $ c $:

- $ f' $ changes **+ → −**: local maximum  
- $ f' $ changes **− → +**: local minimum  
- No sign change: not an extremum

CLEP loves this test because it is conceptual and graphical.

### 4.2 Second Derivative Test
If $ f'(c) = 0 $:

- $ f''(c) > 0 $: local minimum  
- $ f''(c) < 0 $: local maximum  
- $ f''(c) = 0 $: test is inconclusive

CLEP nuance:  
They often give a function where the second derivative test fails and expect you to fall back to the first derivative test.

---

### 5. Curve Sketching Essentials

### 5.1 Increasing/Decreasing Intervals
Use the sign of $ f'(x) $ to determine:
- Where the function rises or falls
- Where local extrema occur

### 5.2 Concavity and Inflection Points
Use the sign of $ f''(x) $ to determine:
- Where the graph bends upward or downward
- Where concavity changes

### 5.3 Asymptotic Behavior
CLEP may ask about:
- Vertical asymptotes (denominator = 0)
- Horizontal asymptotes (limit at infinity)
- End behavior of polynomials

---

### 6. Putting It All Together: CLEP Curve‑Analysis Strategy

To analyze a function on the CLEP exam:

1. **Find the domain**  
   Identify discontinuities.

2. **Compute the first derivative**  
   - Solve $ f'(x) = 0 $ for critical points  
   - Make a sign chart for increasing/decreasing intervals

3. **Compute the second derivative**  
   - Solve $ f''(x) = 0 $ for possible inflection points  
   - Make a concavity sign chart

4. **Classify points**  
   - Use derivative tests to identify maxima/minima  
   - Identify inflection points

5. **Check endpoints** (if on a closed interval)  
   EVT guarantees max/min on closed intervals.

CLEP nuance:  
They often give **graphs of derivatives** and ask you to infer the shape of the original function.

---

### 7. High‑Yield CLEP Question Types

- Identify critical points from a formula or graph  
- Use first derivative test to classify extrema  
- Use second derivative test to classify extrema  
- Determine intervals of increase/decrease  
- Determine intervals of concavity  
- Identify inflection points  
- Interpret graphs of $ f'(x) $ or $ f''(x) $  
- Apply EVT on closed intervals  
- Identify discontinuities and their types  
- Solve optimization problems using derivative tests

---

### 9. CLEP Exam Nuances to Remember

- A limit can exist even if the function value is missing (hole).  
- A function can be continuous but not differentiable (corners, cusps).  
- A derivative graph tells you everything about the shape of the original function.  
- EVT and IVT require continuity; MVT requires differentiability.  
- Second derivative test can fail; first derivative test never fails.  
- Not all critical points are maxima or minima.  
- Inflection points require a **change in concavity**, not just $ f''(x) = 0 $.
