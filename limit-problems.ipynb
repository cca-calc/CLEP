# CLEP-Style Limit Practice by Method
Each problem is paired with a brief solution and explanation. Designed for review, not formal rigor.

## 1. Algebraic Simplification

### 1.1 Factor and Cancel

**Problem 1.1(a)**  
$
\lim_{x \to 3} \frac{x^2 - 9}{x - 3}
$

**Solution:**  
Factor numerator:
$
x^2 - 9 = (x-3)(x+3)
$
So
$
\frac{x^2 - 9}{x - 3} = \frac{(x-3)(x+3)}{x-3} = x+3 \quad (x \ne 3)
$
Now substitute:
$
\lim_{x\to 3} (x+3) = 6
$

**Answer:** $6$  
**Key idea:** Remove the $0/0$ by factoring and canceling.

---

**Problem 1.1(b)**  
$
\lim_{x \to 2} \frac{x^3 - 8}{x - 2}
$

**Solution:**  
Use difference of cubes:
$
x^3 - 8 = (x-2)(x^2 + 2x + 4)
$
So
$
\frac{x^3 - 8}{x - 2} = x^2 + 2x + 4 \quad (x \ne 2)
$
Now substitute:
$
\lim_{x\to 2} (x^2 + 2x + 4) = 4 + 4 + 4 = 12
$

**Answer:** $12$  
**Key idea:** Recognize difference of cubes to eliminate $0/0$.

---

### 1.2 Rationalization (Conjugates)

**Problem 1.2(a)**  
$
\lim_{x \to 0} \frac{\sqrt{x+1} - 1}{x}
$

**Solution:**  
Multiply by conjugate:
$
\frac{\sqrt{x+1} - 1}{x} \cdot \frac{\sqrt{x+1} + 1}{\sqrt{x+1} + 1}
= \frac{(x+1) - 1}{x(\sqrt{x+1} + 1)}
= \frac{x}{x(\sqrt{x+1} + 1)}
= \frac{1}{\sqrt{x+1} + 1}
$
Now substitute:
$
\lim_{x\to 0} \frac{1}{\sqrt{x+1} + 1} = \frac{1}{1+1} = \frac{1}{2}
$

**Answer:** $\frac{1}{2}$  
**Key idea:** Conjugate removes the radical and the $0/0$.

---

**Problem 1.2(b)**  
$
\lim_{x \to 4} \frac{2 - \sqrt{x}}{4 - x}
$

**Solution:**  
Note $4-x = -(x-4)$. Multiply by conjugate of numerator:
$
\frac{2 - \sqrt{x}}{4 - x} \cdot \frac{2 + \sqrt{x}}{2 + \sqrt{x}}
= \frac{4 - x}{(4 - x)(2 + \sqrt{x})}
= \frac{1}{2 + \sqrt{x}}
$
Now substitute:
$
\lim_{x\to 4} \frac{1}{2 + \sqrt{x}} = \frac{1}{2+2} = \frac{1}{4}
$

**Answer:** $\frac{1}{4}$  
**Key idea:** Conjugate converts the expression into a simple rational function.

## 2. Squeeze (Sandwich) Theorem

**Problem 2(a)**  
$
\lim_{x \to 0} x^2 \sin\left(\frac{1}{x}\right)
$

**Solution:**  
We know:
$
-1 \le \sin\left(\frac{1}{x}\right) \le 1
$
Multiply by $x^2 \ge 0$:
$
-x^2 \le x^2 \sin\left(\frac{1}{x}\right) \le x^2
$
As $x \to 0$, both $-x^2$ and $x^2$ go to $0$.  
By Squeeze Theorem:
$
\lim_{x\to 0} x^2 \sin\left(\frac{1}{x}\right) = 0
$

**Answer:** $0$  
**Key idea:** Bound the oscillating part between two functions that both go to 0.

---

**Problem 2(b)**  
$
\lim_{x \to 0} x \cos\left(\frac{1}{x}\right)
$

**Solution:**  
We know:
$
-1 \le \cos\left(\frac{1}{x}\right) \le 1
$
Multiply by $x$ (consider absolute value):
$
-|x| \le x \cos\left(\frac{1}{x}\right) \le |x|
$
As $x \to 0$, both $-|x|$ and $|x|$ go to $0$.  
By Squeeze Theorem:
$
\lim_{x\to 0} x \cos\left(\frac{1}{x}\right) = 0
$

**Answer:** $0$  
**Key idea:** Use absolute value to control sign and squeeze to 0.

## 3. L'Hôpital's Rule

### 3.1 Direct 0/0 or ∞/∞

**Problem 3.1(a)**  
$
\lim_{x \to 0} \frac{\sin x}{x}
$

**Solution (L'Hôpital):**  
As $x \to 0$, $\sin x \to 0$ and $x \to 0$ → $0/0$.  
Differentiate top and bottom:
$
\lim_{x\to 0} \frac{\sin x}{x}
= \lim_{x\to 0} \frac{\cos x}{1} = \cos 0 = 1
$

**Answer:** $1$  
**Key idea:** Classic 0/0 form; derivative of $\sin x$ is $\cos x$.

---

**Problem 3.1(b)**  
$
\lim_{x \to \infty} \frac{\ln x}{x}
$

**Solution:**  
As $x \to \infty$, $\ln x \to \infty$ and $x \to \infty$ → $\infty/\infty$.  
Apply L'Hôpital:
$
\lim_{x\to\infty} \frac{\ln x}{x}
= \lim_{x\to\infty} \frac{1/x}{1}
= \lim_{x\to\infty} \frac{1}{x} = 0
$

**Answer:** $0$  
**Key idea:** Polynomial growth dominates logarithmic growth.

---

### 3.2 Converted Indeterminate Forms

**Problem 3.2(a)**  ($0 \cdot \infty$ type)  
$
\lim_{x \to 0^+} x \ln x
$

**Solution:**  
Rewrite as a quotient:
$
x \ln x = \frac{\ln x}{1/x}
$
As $x \to 0^+$, $\ln x \to -\infty$, $1/x \to \infty$ → $-\infty/\infty$ (indeterminate).  
Apply L'Hôpital:
$
\lim_{x\to 0^+} \frac{\ln x}{1/x}
= \lim_{x\to 0^+} \frac{1/x}{-1/x^2}
= \lim_{x\to 0^+} \left( \frac{1}{x} \cdot \frac{x^2}{-1} \right)
= \lim_{x\to 0^+} (-x) = 0
$

**Answer:** $0$  
**Key idea:** Convert product to quotient, then use L'Hôpital.

---

**Problem 3.2(b)**  ($1^\infty$ type)  
$
\lim_{x \to \infty} \left(1 + \frac{1}{x}\right)^x
$

**Solution:**  
Let
$
y = \left(1 + \frac{1}{x}\right)^x
$
Take natural log:
$
\ln y = x \ln\left(1 + \frac{1}{x}\right)
$
Now consider:
$
\lim_{x\to\infty} x \ln\left(1 + \frac{1}{x}\right)
= \lim_{x\to\infty} \frac{\ln\left(1 + \frac{1}{x}\right)}{1/x}
$
As $x\to\infty$, numerator $\to 0$, denominator $\to 0$ → $0/0$.  
Apply L'Hôpital:
$
\lim_{x\to\infty} \frac{\ln\left(1 + \frac{1}{x}\right)}{1/x}
= \lim_{x\to\infty} \frac{\frac{1}{1+1/x} \cdot \left(-\frac{1}{x^2}\right)}{-1/x^2}
= \lim_{x\to\infty} \frac{1}{1+1/x} = 1
$
So
$
\lim_{x\to\infty} \ln y = 1 \Rightarrow \lim_{x\to\infty} y = e^1 = e
$

**Answer:** $e$  
**Key idea:** Log trick converts $1^\infty$ to a product suitable for L'Hôpital.

## 4. Dominant Term / End Behavior

**Problem 4(a)**  
$
\lim_{x \to \infty} \frac{3x^2 - x + 1}{2x^2 + 5}
$

**Solution:**  
Compare leading terms:
$
\frac{3x^2 - x + 1}{2x^2 + 5} \approx \frac{3x^2}{2x^2} = \frac{3}{2}
$
Formally, divide numerator and denominator by $x^2$:
$
\lim_{x\to\infty} \frac{3 - 1/x + 1/x^2}{2 + 5/x^2} = \frac{3}{2}
$

**Answer:** $\frac{3}{2}$  
**Key idea:** Same degree → ratio of leading coefficients.

---

**Problem 4(b)**  
$
\lim_{x \to \infty} \frac{5x - 7}{x^2 + 1}
$

**Solution:**  
Degree of denominator is higher. Divide by $x^2$:
$
\frac{5x - 7}{x^2 + 1} = \frac{5/x - 7/x^2}{1 + 1/x^2}
$
As $x\to\infty$, all terms with $1/x$ or $1/x^2$ go to 0:
$
\lim_{x\to\infty} \frac{5/x - 7/x^2}{1 + 1/x^2} = 0
$

**Answer:** $0$  
**Key idea:** Denominator grows faster → limit is 0.

## 5. Special Trig Limits / Small-Angle Approximations

**Problem 5(a)**  
$
\lim_{x \to 0} \frac{\sin(3x)}{x}
$

**Solution:**  
Rewrite:
$
\frac{\sin(3x)}{x} = 3 \cdot \frac{\sin(3x)}{3x}
$
We know:
$
\lim_{u\to 0} \frac{\sin u}{u} = 1
$
Let $u = 3x$. Then as $x\to 0$, $u\to 0$:
$
\lim_{x\to 0} \frac{\sin(3x)}{3x} = 1
$
So:
$
\lim_{x\to 0} \frac{\sin(3x)}{x} = 3 \cdot 1 = 3
$

**Answer:** $3$  
**Key idea:** Factor out the coefficient inside the sine.

---

**Problem 5(b)**  
$
\lim_{x \to 0} \frac{1 - \cos x}{x^2}
$

**Solution:**  
Known standard limit:
$
\lim_{x\to 0} \frac{1 - \cos x}{x^2} = \frac{1}{2}
$
One way to see it: use identity $1 - \cos x = 2\sin^2(x/2)$:
$
\frac{1 - \cos x}{x^2} = \frac{2\sin^2(x/2)}{x^2}
= 2 \left(\frac{\sin(x/2)}{x/2}\right)^2 \cdot \frac{1}{4}
= \frac{1}{2} \left(\frac{\sin(x/2)}{x/2}\right)^2
$
As $x\to 0$, $\frac{\sin(x/2)}{x/2} \to 1$, so the limit is $\frac{1}{2}$.

**Answer:** $\frac{1}{2}$  
**Key idea:** Use trig identity plus the basic $\sin u / u$ limit.

## 6. Numerical / Graphical Intuition (Conceptual)

**Problem 6(a)**  
Given that $f(x) = \frac{|x|}{x}$, evaluate:
$
\lim_{x \to 0^-} f(x), \quad \lim_{x \to 0^+} f(x), \quad \lim_{x \to 0} f(x)
$

**Solution:**  
For $x < 0$, $|x| = -x$, so $f(x) = \frac{-x}{x} = -1$.  
For $x > 0$, $|x| = x$, so $f(x) = \frac{x}{x} = 1$.

So:
$
\lim_{x\to 0^-} f(x) = -1, \quad \lim_{x\to 0^+} f(x) = 1
$
Since left and right limits differ, the two-sided limit does not exist.

**Answer:**  
$\lim_{x\to 0^-} f(x) = -1$,  
$\lim_{x\to 0^+} f(x) = 1$,  
$\lim_{x\to 0} f(x)$ does not exist.

**Key idea:** Jump discontinuity; one-sided limits differ.

---

**Problem 6(b)**  
Let
$
g(x) = 
\begin{cases}
x^2, & x \ne 2 \\
5, & x = 2
\end{cases}
$
Find $\lim_{x \to 2} g(x)$.

**Solution:**  
For $x \ne 2$, $g(x) = x^2$, which is continuous.  
So:
$
\lim_{x\to 2} g(x) = \lim_{x\to 2} x^2 = 4
$
Note that $g(2) = 5$, but that does not affect the limit.

**Answer:** $4$  
**Key idea:** Limit depends on nearby values, not the value at the point.
