# Limit-Solving Methods for CLEP Calculus
A structured reference for algebraic, analytic, and theorem-based limit techniques.

---

## 1. Algebraic Simplification Methods

### 1.1 Factor and Cancel
- Factor numerator/denominator
- Cancel common factors to remove 0/0 form  
**Examples:**  
- $\lim_{x\to a} \frac{x^2 - a^2}{x-a}$  
- $\lim_{x\to 2} \frac{x^3 - 8}{x-2}$

---

### 1.2 Rationalization (Conjugates)
Used when radicals produce 0/0.  
Multiply by conjugate to eliminate square roots.  
**Examples:**  
- $\lim_{x\to 0} \frac{\sqrt{x+1}-1}{x}$  
- $\lim_{x\to 4} \frac{2 - \sqrt{x}}{4-x}$

---

### 1.3 Common Denominator / Combine Expressions
Used for expressions like $\infty - \infty$ or sums of fractions.  
**Examples:**  
- $\lim_{x\to\infty} \left(\sqrt{x^2+1} - x\right)$  
- $\lim_{x\to 0} \left(\frac{1}{x} - \frac{1}{x+1}\right)$

---

### 1.4 Polynomial / Rational Dominance (End Behavior)
Use leading-term behavior for limits at infinity.  
**Rules:**  
- Degree(top) < Degree(bottom) → limit = 0  
- Degree(top) = Degree(bottom) → ratio of leading coefficients  
- Degree(top) > Degree(bottom) → ±∞  

---

### 1.5 Special Trig Limits
Memorize these:
- $\lim_{x\to 0} \frac{\sin x}{x} = 1$  
- $\lim_{x\to 0} \frac{1 - \cos x}{x} = 0$  
- $\lim_{x\to 0} \frac{\tan x}{x} = 1$

---

## 2. Theorems for Evaluating Limits

### 2.1 Squeeze (Sandwich) Theorem
Used when a function is trapped between two simpler functions.  
**Typical patterns:**  
- $\sin x \le x \le \tan x$ near 0  
- $|\sin(1/x)| \le 1$  
- $|x|\sin(1/x) \to 0$

---

### 2.2 Limit Laws
Break complicated limits into sums, products, quotients, powers.  
**Examples:**  
- $\lim (f+g) = \lim f + \lim g$  
- $\lim (cf) = c \lim f$  
- $\lim (fg) = (\lim f)(\lim g)$

---

### 2.3 Continuity Rules
If a function is continuous at a point, evaluate by substitution.  
**Examples:**  
- Polynomials  
- Rational functions (where denominator ≠ 0)  
- Exponential, log, trig, inverse trig

---
### 3. L'Hôpital's Rule

- Evaluate limits using L'Hôpital's rule (methods to fit expressions into $\frac{0}{0}$ or $\frac{\infty}{\infty}$).

<div align="center">

| Indeterminate Form | How to Rewrite | Example Setup | Result after Rewriting |
|--------------------|----------------|---------------|------------------------|
| $\tfrac{0}{0} $ | Already in quotient form | $lim_{x \to 0} \frac{\sin x}{x}$ | Apply L'Hôpital directly |
| $\tfrac{\infty}{\infty} $ | Already in quotient form | $\lim_{x \to \infty} \frac{x}{e^x}$ | Apply L'Hôpital directly |
| $0 \cdot \infty$ | Rewrite as $\frac{\ln x}{1/x}$ or similar | $\lim_{x \to 0^+} x \ln x$ | Becomes $\tfrac{\infty}{\infty}$ |
| $\infty - \infty$ | Combine into single fraction (e.g. conjugate) | $\lim_{x \to \infty} (\sqrt{x^2+1} - x)$ | Becomes $\tfrac{0}{0}$ |
| $0^0$ | Take logs: $g(x)\ln f(x)$ | $\lim_{x \to 0^+} x^x$ | Reduces to $0 \cdot \infty$ |
| $\infty^0$ | Take logs: $g(x)\ln f(x)$ | $\lim_{x \to \infty} (1+1/x)^x$ | Reduces to $\tfrac{0}{0}$ |
| $1^\infty$ | Take logs: $g(x)\ln f(x)$ | $\lim_{x \to \infty} (1+1/x)^x$ | Reduces to $\tfrac{0}{0}$ |

</div>

## 4. Numerical / Graphical Reasoning (CLEP-friendly)
- Evaluate values approaching from left/right  
- Recognize jump discontinuities  
- Identify vertical/horizontal asymptotes  
- Use tables or graph behavior
