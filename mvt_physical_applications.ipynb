# Applications of the Mean Value Theorem to Physical Problems

This notebook summarizes how the Mean Value Theorem (MVT) appears in physics and physical modeling. Each section includes the core idea, the MVT connection, and a physical interpretation.

The Mean Value Theorem states:

$$
f'(c) = \frac{f(b)-f(a)}{b-a}
$$
for some $c$ in $(a,b)$, provided $f$ is continuous on $[a,b]$ and differentiable on $(a,b)$.


## 1. Motion Along a Line: Instantaneous Velocity Equals Average Velocity

**Physical quantity:** position $s(t)$

**Average velocity:**
$$
\frac{s(b)-s(a)}{b-a}
$$

**MVT conclusion:** There exists a time $c$ such that
$$
s'(c) = \frac{s(b)-s(a)}{b-a}.
$$

**Interpretation:** At some moment, the instantaneous velocity equals the average velocity over the time interval.

**Example:** If a car travels 100 miles in 2 hours, then at some instant the speedometer must read exactly 50 mph.


## 2. Bounding How Fast a Quantity Can Change

If a physical quantity $f(t)$ has a known bound on its rate of change, such as
$$
|f'(t)| \le M,
$$
then the MVT implies
$$
|f(b)-f(a)| \le M|b-a|.
$$

**Interpretation:** A system with limited acceleration, heating rate, or force cannot change too quickly.

**Examples:**
- A spacecraft with maximum acceleration $A$ cannot change velocity by more than $A\Delta t$.
- A metal rod with bounded heating rate cannot exceed certain temperature differences over short times.


## 3. Fluid Flow and Average Flow Rate

Let $Q(t)$ be the volume of fluid that has passed through a pipe by time $t$.

**Average flow rate:**
$$
\frac{Q(b)-Q(a)}{b-a}
$$

By the MVT, there exists $c$ such that
$$
Q'(c) = \frac{Q(b)-Q(a)}{b-a}.
$$

**Interpretation:** At some moment, the instantaneous flow rate equals the average flow rate over the interval.

**Physical meaning:** Even if flow is turbulent or oscillating, there must be a moment where the flow matches the average throughput.


## 4. Power and Work: Instantaneous Power Equals Average Power

Let $W(t)$ be the work done by a force up to time $t$.

**Average power:**
$$
\frac{W(b)-W(a)}{b-a}
$$

By the MVT, there exists $c$ such that
$$
W'(c) = \frac{W(b)-W(a)}{b-a}.
$$

But $W'(t)$ is instantaneous power $P(t)$, so
$$
P(c) = \text{average power over } [a,b].
$$

**Interpretation:** During any interval of work, there is a moment when the instantaneous power output equals the average power.

$$
T'(c) = \frac{T(b)-T(a)}{b-a}.

**Applications:**

## 6. Population Growth and Average Growth Rate

Let $P(t)$ be the population of a species or system.

**Average growth rate:**
$$
\frac{P(b)-P(a)}{b-a}
$$

By the MVT, there exists $c$ such that

P'(c) = \frac{P(b)-P(a)}{b-a}.

## 7. Electrical Charge and Current

Let $Q(t)$ be electric charge.

**Average current:**
$$
\frac{Q(b)-Q(a)}{b-a}
$$

Since current is $I(t)=Q'(t)$, the MVT gives
$$
I(c) = \frac{Q(b)-Q(a)}{b-a}.
$$

**Interpretation:** At some instant, the instantaneous current equals the average current over the interval.

**Applications:**
- Circuit analysis
- Battery discharge modeling
- Signal processing

|-----------------|----------------|--------------------|

The Mean Value Theorem provides a bridge between average behavior and instantaneous behavior in any physical system where a quantity changes smoothly over time.
