# Problem Set: Average Rate of Change, Mean Value Theorem, and Net Change Theorem (Chat GPT 5.1)

## Problems

### 1. Polynomial motion
Let $s(t) = t^3 - 6t^2 + 9t$ be the position of a particle on $[0,4]$.

1. Find the average velocity on $[0,4]$.
2. Use the MVT to find all $c\in(0,4)$ where the instantaneous velocity equals the average velocity.

---

### 2. Net change from velocity
For the same $s(t)$ as in Problem 1, compute the net displacement from $t=0$ to $t=4$ using the velocity $v(t)=s'(t)$ and the Net Change Theorem.

---

### 3. Quadratic function
Let $F(x) = x^2 - 4x + 7$ on $[1,5]$.

1. Compute the average rate of change of $F$ on $[1,5]$.
2. Find $c$ in $(1,5)$ such that $F'(c)$ equals this average rate (MVT).

---

### 4. Connecting AROC and integral
For $F(x)$ in Problem 3, verify that
$
\frac{F(5)-F(1)}{5-1} = \frac{1}{5-1}\int_1^5 F'(x)\,dx.
$

3. Use the MVT to find $c\in(0,5)$ such that $P'(c)$ equals this average rate.

# Worked Solutions

### 1.
$s(t) = t^3 - 6t^2 + 9t$.

1. $s(0)=0$, $s(4)=64 - 96 + 36 = 4$.  
   $\text{AROC} = \dfrac{4-0}{4-0} = 1.$

2. $v(t)=s'(t)=3t^2 - 12t + 9.$  
   Set $v(c)=1$:  
   $3c^2 - 12c + 9 = 1 \Rightarrow 3c^2 - 12c + 8 = 0 \Rightarrow c^2 - 4c + \tfrac{8}{3}=0.$  
   Discriminant $=16 - \tfrac{32}{3} = \tfrac{16}{3}$.  
   $c = \dfrac{4 \pm \sqrt{16/3}}{2} = 2 \pm \dfrac{2}{\sqrt{3}}.$  
   Both are in $(0,4)$.

---

### 2.
Net displacement:
$
\int_0^4 v(t)\,dt = s(4) - s(0) = 4 - 0 = 4.
$

---

### 3.
$F(x) = x^2 - 4x + 7$ on $[1,5]$.

1. $F(1)=1-4+7=4$, $F(5)=25-20+7=12.$  
   $\text{AROC} = \dfrac{12-4}{5-1} = \dfrac{8}{4} = 2.$

2. $F'(x)=2x-4.$  
   Set $2c - 4 = 2 \Rightarrow 2c = 6 \Rightarrow c = 3.$

---

### 4.
$F'(x)=2x-4.$

$
\int_1^5 F'(x)\,dx = \int_1^5 (2x-4)\,dx
= \left[x^2 - 4x\right]_1^5
= (25 - 20) - (1 - 4) = 5 - (-3) = 8.
$

Then
$
\frac{1}{5-1}\int_1^5 F'(x)\,dx = \frac{1}{4}\cdot 8 = 2,
$
which matches $\dfrac{F(5)-F(1)}{4}$.

---

### 5.
$P'(t) = 200e^{0.1t}$.

1. Net change:
$
\int_0^5 200e^{0.1t}\,dt
= 200\cdot \frac{1}{0.1}\left[e^{0.1t}\right]_0^5
= 2000(e^{0.5} - 1).
$

2. Average rate:
$
\text{AROC} = \frac{2000(e^{0.5}-1)}{5} = 400(e^{0.5}-1).
$

3. MVT: solve $P'(c) = 400(e^{0.5}-1)$:
$
200e^{0.1c} = 400(e^{0.5}-1)
\Rightarrow e^{0.1c} = 2(e^{0.5}-1)
\Rightarrow c = 10\ln\big(2(e^{0.5}-1)\big).
$

---

### 6.
$v(t) = 3 + 2\sin t$ on $[0,2\pi]$.

1. Net displacement:
$
\int_0^{2\pi} (3 + 2\sin t)\,dt
= \left[3t - 2\cos t\right]_0^{2\pi}
= (6\pi - 2\cos 2\pi) - (0 - 2\cos 0)
= 6\pi - 2 + 2 = 6\pi.
$

2. Average velocity:
$
\text{AROC} = \frac{6\pi}{2\pi} = 3.
$

3. Solve $v(c)=3$:  
$3 + 2\sin c = 3 \Rightarrow \sin c = 0.$  
On $(0,2\pi)$: $c = \pi.$

---

### 7.
$f(x) = 5x - 3$ on $[2,7]$.

1. $f(2)=7$, $f(7)=32.$  
   $\text{AROC} = \dfrac{32-7}{7-2} = \dfrac{25}{5} = 5.$

2. $f'(x)=5$. So $f'(c)=5$ for all $c$, so every point satisfies the MVT condition.

3. This makes sense because the graph is a straight line; the secant slope equals the tangent slope everywhere.

---

### 8.
$r(t) = t$ on $[0,2]$, $r(t)=2$ on $(2,5]$.

1. Net change:
$
\int_0^2 t\,dt + \int_2^5 2\,dt
= \left[\frac{t^2}{2}\right]_0^2 + \left[2t\right]_2^5
= 2 + (10 - 4) = 2 + 6 = 8.
$

2. Average rate over $[0,5]$:
$
\text{AROC} = \frac{8}{5-0} = \frac{8}{5}.
$

---

### 9.
$f(x)=\sin x$, $F(x)=\int_0^x \sin t\,dt$.

1. $F(b)-F(a) = F(\pi)-F(0)$.  
   $F(x) = 1 - \cos x$.  
   So $F(\pi)-F(0) = (1 - \cos\pi) - (1 - \cos 0) = (1 - (-1)) - (1 - 1) = 2.$

2. Average rate on $[0,\pi]$:
$
\text{AROC} = \frac{2}{\pi}.
$

3. MVT: $F'(x)=\sin x$.  
   Solve $\sin c = \dfrac{2}{\pi}$ for $c\in(0,\pi)$.  
   So $c = \arcsin\left(\dfrac{2}{\pi}\right)$.  
   Interpretation: at $x=c$, the instantaneous rate $F'(c)=\sin c$ equals the average rate of change of $F$ on $[0,\pi]$.

---

### 10.
$T'(t) = -4e^{-0.5t}$ on $[0,6]$.

1. Net change:
$
\int_0^6 -4e^{-0.5t}\,dt
= -4\cdot \left[\frac{1}{-0.5}e^{-0.5t}\right]_0^6
= 8(e^{-3} - 1).
$

2. Average rate:
$
\text{AROC} = \frac{8(e^{-3}-1)}{6} = \frac{4}{3}(e^{-3}-1).
$

3. MVT: solve $T'(c) = \dfrac{4}{3}(e^{-3}-1)$:
$
-4e^{-0.5c} = \frac{4}{3}(e^{-3}-1)
\Rightarrow e^{-0.5c} = -\frac{1}{3}(e^{-3}-1) = \frac{1 - e^{-3}}{3}.
$
So
$
c = -2\ln\left(\frac{1 - e^{-3}}{3}\right).
$

---

### 11.
$F(x) = x^3 - 3x^2 + 2x$ on $[0,3]$.

1. $F(0)=0$, $F(3)=27 - 27 + 6 = 6.$  
   $\text{AROC} = \dfrac{6-0}{3-0} = 2.$

2. $F'(x)=3x^2 - 6x + 2.$  
   Solve $3c^2 - 6c + 2 = 2$:  
   $3c^2 - 6c = 0 \Rightarrow 3c(c-2)=0 \Rightarrow c=0$ or $c=2$.  
   In $(0,3)$, $c=2$.

---

### 12.
Verify:
$
\frac{F(3)-F(0)}{3} = \frac{1}{3}\int_0^3 F'(x)\,dx.
$

Left side: $\dfrac{6}{3} = 2.$

Right side:
$
\int_0^3 (3x^2 - 6x + 2)\,dx
= \left[x^3 - 3x^2 + 2x\right]_0^3
= 6.
$
Then $\dfrac{1}{3}\cdot 6 = 2.$  
So the average rate of change of $F$ equals the average value of $F'$.

---

### 13.
$q(t) = 10 + 4t$ on $[0,5]$.

1. Total volume:
$
\int_0^5 (10 + 4t)\,dt
= \left[10t + 2t^2\right]_0^5
= 50 + 50 = 100.
$

2. Average flow rate:
$
\text{AROC} = \frac{100}{5} = 20.
$

3. Solve $q(c)=20$:  
$10 + 4c = 20 \Rightarrow 4c = 10 \Rightarrow c = 2.5.$

---

### 14.
$G'(x) = 2x - 1$, $G(0)=3$.

1. Net change:
$
G(4)-G(0) = \int_0^4 (2x - 1)\,dx
= \left[x^2 - x\right]_0^4
= (16 - 4) - 0 = 12.
$

2. Average rate:
$
\text{AROC} = \frac{12}{4} = 3.
$

3. MVT: solve $G'(c)=3$:  
$2c - 1 = 3 \Rightarrow 2c = 4 \Rightarrow c = 2.$

---

### 15.
$H'(x) = e^x$ on $[0,2]$.

1. Net change:
$
H(2)-H(0) = \int_0^2 e^x\,dx = e^2 - 1.
$

2. Average rate:
$
\text{AROC} = \frac{e^2 - 1}{2}.
$

3. MVT: solve $H'(c) = e^c = \dfrac{e^2 - 1}{2}$.  
So
$
c = \ln\left(\frac{e^2 - 1}{2}\right),
$
which lies in $(0,2)$.
