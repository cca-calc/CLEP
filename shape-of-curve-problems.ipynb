### CLEP‑Style Problem Set: Functions, Curve Shape, Critical Points, Extrema, Derivative Tests

---

### Problem 1 — Critical Points and Local Extrema

Let
$
f(x) = x^3 - 3x^2 + 2.
$

1. Find all critical points of $ f $.  
2. Use the **first derivative test** to classify each critical point as a local maximum, local minimum, or neither.  
3. Determine the intervals where $ f $ is increasing and decreasing.

---

### Problem 2 — Second Derivative Test

Let
$
f(x) = x^4 - 4x^2.
$

1. Find all critical points.  
2. Compute $ f''(x) $.  
3. Use the **second derivative test** to classify each critical point as a local max, local min, or inconclusive.  
4. Confirm your classification using the sign of $ f'(x) $ if needed.

---

### Problem 3 — Concavity and Inflection Points

Let
$
f(x) = x^3 - 6x.
$

1. Compute $ f'(x) $ and $ f''(x) $.  
2. Find all intervals where $ f $ is concave up and concave down.  
3. Find all inflection points (if any).

---

### Problem 4 — Interpreting the Derivative Graph

Suppose the graph of $ f'(x) $ is given (not $ f(x) $) and is described as follows:

- $ f'(x) > 0 $ on $ (-\infty, -1) $  
- $ f'(x) < 0 $ on $ (-1, 2) $  
- $ f'(x) > 0 $ on $ (2, \infty) $

1. On which intervals is $ f $ increasing? Decreasing?  
2. Classify $ x = -1 $ and $ x = 2 $ as local maxima, minima, or neither.  
3. Sketch a rough shape of $ f(x) $ consistent with this information.

---

### Problem 5 — Mixed First and Second Derivative Information

Let
$
f'(x) = (x-1)(x+2), \quad f''(x) = 2x+1.
$

1. Find the critical points of $ f $.  
2. Use the sign of $ f'(x) $ to determine where $ f $ is increasing and decreasing.  
3. Use $ f''(x) $ to determine concavity and any inflection points.  
4. Classify each critical point as a local max, local min, or neither.

### Problem 6 — Endpoints and Absolute Extrema (EVT)

Let
$
f(x) = x^3 - 3x
$
on the closed interval $[-2, 2]$.

1. Find all critical points in $[-2, 2]$.  
2. Evaluate $ f $ at the critical points and at the endpoints.  
3. Determine the **absolute maximum** and **absolute minimum** of $ f $ on $[-2, 2]$.

# Solutions: CLEP‑Style Problem Set

---

### Problem 1 — Critical Points and Local Extrema

$
f(x) = x^3 - 3x^2 + 2
$

**1. Critical points**

$
f'(x) = 3x^2 - 6x = 3x(x-2)
$

Critical points where $ f'(x) = 0 $:

$
x = 0,\; x = 2
$

**2. First derivative test**

Sign of $ f'(x) = 3x(x-2) $:

- For $ x < 0 $: $ x < 0,\; x-2 < 0 \Rightarrow f'(x) > 0 $ (increasing)
- For $ 0 < x < 2 $: $ x > 0,\; x-2 < 0 \Rightarrow f'(x) < 0 $ (decreasing)
- For $ x > 2 $: $ x > 0,\; x-2 > 0 \Rightarrow f'(x) > 0 $ (increasing)

So:
- At $ x = 0 $: $ f' $ changes $ + \to - $ → **local maximum**  
- At $ x = 2 $: $ f' $ changes $ - \to + $ → **local minimum**

**3. Increasing/decreasing intervals**

- Increasing on $ (-\infty, 0) $ and $ (2, \infty) $  
- Decreasing on $ (0, 2) $

---

### Problem 2 — Second Derivative Test

$
f(x) = x^4 - 4x^2
$

**1. Critical points**

$
f'(x) = 4x^3 - 8x = 4x(x^2 - 2)
$

Set $ f'(x) = 0 $:

$
4x(x^2 - 2) = 0 \Rightarrow x = 0,\; x = \pm \sqrt{2}
$

**2. Second derivative**

$
f''(x) = 12x^2 - 8
$

**3. Second derivative test**

- At $ x = 0 $:  
  $
  f''(0) = -8 < 0 \Rightarrow \text{local maximum}
  $
- At $ x = \sqrt{2} $:  
  $
  f''(\sqrt{2}) = 12(2) - 8 = 16 > 0 \Rightarrow \text{local minimum}
  $
- At $ x = -\sqrt{2} $:  
  Same value $ 16 > 0 \Rightarrow \text{local minimum} $

**4. First derivative sign (confirmation)**  
Matches the above classification.

---

### Problem 3 — Concavity and Inflection Points

$
f(x) = x^3 - 6x
$

**1. Derivatives**

$
f'(x) = 3x^2 - 6,\quad f''(x) = 6x
$

**2. Concavity**

Set $ f''(x) = 0 \Rightarrow 6x = 0 \Rightarrow x = 0 $.

- For $ x < 0 $: $ f''(x) < 0 \Rightarrow $ concave down  
- For $ x > 0 $: $ f''(x) > 0 \Rightarrow $ concave up  

So:
- Concave down on $ (-\infty, 0) $  
- Concave up on $ (0, \infty) $

**3. Inflection point**

Concavity changes at $ x = 0 $.  
$
f(0) = 0
$
Inflection point at $ (0, 0) $.

### Problem 4 — Interpreting the Derivative Graph

Given:

- $ f'(x) > 0 $ on $ (-\infty, -1) $  
- $ f'(x) < 0 $ on $ (-1, 2) $  
- $ f'(x) > 0 $ on $ (2, \infty) $

**1. Increasing/decreasing**

- Increasing on $ (-\infty, -1) $ and $ (2, \infty) $  
- Decreasing on $ (-1, 2) $

**2. Classify $ x = -1, 2 $**

- At $ x = -1 $: $ f' $ changes $ + \to - $ → **local maximum**  
- At $ x = 2 $: $ f' $ changes $ - \to + $ → **local minimum**

**3. Rough shape**

- Rises to a peak at $ x = -1 $, then falls to a valley at $ x = 2 $, then rises again.

### Problem 5 — Mixed First and Second Derivative Information

$
f'(x) = (x-1)(x+2),\quad f''(x) = 2x + 1
$

**1. Critical points**

Set $ f'(x) = 0 $:

$
(x-1)(x+2) = 0 \Rightarrow x = 1,\; x = -2
$

**2. Increasing/decreasing**

Sign of $ f'(x) $:

- For $ x < -2 $: both factors negative → $ f'(x) > 0 $ (increasing)  
- For $ -2 < x < 1 $: one negative, one positive → $ f'(x) < 0 $ (decreasing)  
- For $ x > 1 $: both positive → $ f'(x) > 0 $ (increasing)

So:
- Increasing on $ (-\infty, -2) $ and $ (1, \infty) $  
- Decreasing on $ (-2, 1) $

**3. Concavity and inflection**

Set $ f''(x) = 0 \Rightarrow 2x + 1 = 0 \Rightarrow x = -\tfrac{1}{2} $.

- For $ x < -\tfrac{1}{2} $: $ f''(x) < 0 $ → concave down  
- For $ x > -\tfrac{1}{2} $: $ f''(x) > 0 $ → concave up  

Inflection point at $ x = -\tfrac{1}{2} $ (actual $ f(x) $ value not needed for CLEP‑style conceptual questions).

**4. Classify critical points**

Use $ f''(x) $:

- At $ x = -2 $:  
  $
  f''(-2) = 2(-2) + 1 = -3 < 0 \Rightarrow \text{local maximum}
  $
- At $ x = 1 $:  
  $
  f''(1) = 2(1) + 1 = 3 > 0 \Rightarrow \text{local minimum}
  $

### Problem 6 — Endpoints and Absolute Extrema (EVT)

$
f(x) = x^3 - 3x,\quad x \in [-2, 2]
$

**1. Critical points**

$
f'(x) = 3x^2 - 3 = 3(x^2 - 1)
$

Set $ f'(x) = 0 \Rightarrow x^2 - 1 = 0 \Rightarrow x = \pm 1 $.

Both are in $[-2, 2]$.

**2. Evaluate at critical points and endpoints**

$
f(-2) = (-2)^3 - 3(-2) = -8 + 6 = -2
$
$
f(-1) = (-1)^3 - 3(-1) = -1 + 3 = 2
$
$
f(1) = 1^3 - 3(1) = 1 - 3 = -2
$
$
f(2) = 8 - 6 = 2
$

**3. Absolute extrema**

- Maximum value: $ 2 $ at $ x = -1 $ and $ x = 2 $  
- Minimum value: $ -2 $ at $ x = -2 $ and $ x = 1 $

### Problem 7 — Optimization Style

Rectangle with base on $ x $-axis, top corners on $ y = 8 - x^2 $, $ x \ge 0 $.

Let the right top corner be at $ (x, 8 - x^2) $. By symmetry, left corner is at $ (-x, 8 - x^2) $.

**1. Area function**

Width $ = 2x $, height $ = 8 - x^2 $.

$
A(x) = 2x(8 - x^2) = 16x - 2x^3
$

Domain: $ 8 - x^2 \ge 0 \Rightarrow x^2 \le 8 \Rightarrow 0 \le x \le \sqrt{8} $.

**2. Critical points**

$
A'(x) = 16 - 6x^2
$

Set $ A'(x) = 0 $:

$
16 - 6x^2 = 0 \Rightarrow 6x^2 = 16 \Rightarrow x^2 = \frac{8}{3} \Rightarrow x = \sqrt{\frac{8}{3}}
$

(Only positive root relevant.)

**3. Maximum area**

Check that this is a maximum (e.g., $ A''(x) = -12x < 0 $ for $ x > 0 $).

$
A\left(\sqrt{\frac{8}{3}}\right)
= 2\sqrt{\frac{8}{3}}\left(8 - \frac{8}{3}\right)
= 2\sqrt{\frac{8}{3}} \cdot \frac{16}{3}
= \frac{32}{3}\sqrt{\frac{8}{3}}
$

That is the maximum area (exact form is fine for CLEP‑style).

### Problem 8 — Concavity From Second Derivative Only

$
f''(x) = x^2 - 4
$

**1. Concavity intervals**

Set $ f''(x) = 0 \Rightarrow x^2 - 4 = 0 \Rightarrow x = \pm 2 $.

- For $ x < -2 $: $ x^2 - 4 > 0 \Rightarrow $ concave up  
- For $ -2 < x < 2 $: $ x^2 - 4 < 0 \Rightarrow $ concave down  
- For $ x > 2 $: $ x^2 - 4 > 0 \Rightarrow $ concave up  

So:
- Concave up on $ (-\infty, -2) \cup (2, \infty) $  
- Concave down on $ (-2, 2) $

**2. Possible inflection points**

Concavity changes at $ x = -2 $ and $ x = 2 $, so these are **candidates**.

**3. Additional information**

To confirm actual inflection points, we need to know that $ f $ is defined and differentiable at those $ x $-values and that concavity indeed changes (which we already see from sign change of $ f'' $). In a full problem, we’d also evaluate $ f(x) $ at those points.

### Problem 9 — Inconclusive Second Derivative Test

$
f(x) = x^3
$

**1. Critical points**

$
f'(x) = 3x^2
$

Set $ f'(x) = 0 \Rightarrow 3x^2 = 0 \Rightarrow x = 0 $.

**2. Second derivative**

$
f''(x) = 6x,\quad f''(0) = 0
$

Second derivative test is **inconclusive** when $ f''(c) = 0 $.

**3. Why inconclusive**

The test only works when $ f''(c) \neq 0 $. Here, $ f''(0) = 0 $, so we cannot classify using this test.

**4. First derivative test / reasoning**

Check sign of $ f'(x) = 3x^2 $:

- For $ x < 0 $: $ f'(x) > 0 $  
- For $ x > 0 $: $ f'(x) > 0 $

No sign change; function is increasing on both sides.  
So $ x = 0 $ is **not** a local max or min; it is a point of inflection with horizontal tangent.

### Problem 10 — Multiple‑Choice CLEP Style

$
f(x) = x^3 - 6x^2 + 9x
$

**1. First derivative**

$
f'(x) = 3x^2 - 12x + 9 = 3(x^2 - 4x + 3) = 3(x-1)(x-3)
$

Critical points: $ x = 1,\; x = 3 $.

**2. Second derivative**

$
f''(x) = 6x - 12
$

- At $ x = 1 $: $ f''(1) = 6 - 12 = -6 < 0 \Rightarrow $ local maximum  
- At $ x = 3 $: $ f''(3) = 18 - 12 = 6 > 0 \Rightarrow $ local minimum  

So $ f $ has **one local maximum and one local minimum**.

Correct answer: **C**.
